```python
# ==========================================================
# FULL PIPELINE: mount Drive -> unzip dataset -> train -> eval -> save models
# ==========================================================
import os, sys, shutil, random, zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.applications.densenet import preprocess_input
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization, Input
from tensorflow.keras import regularizers, optimizers

print("âœ… Imports done")

# ---------------------------
# 0) MOUNT GOOGLE DRIVE (Colab)
# ---------------------------
try:
    from google.colab import drive
    drive.mount('/content/drive', force_remount=False)
    print("âœ… Google Drive mounted at /content/drive")
except Exception as e:
    print("âš ï¸ Not running inside Colab or google.colab unavailable. Ensure ZIP path is accessible.")
    # continue â€” user may run locally and provide correct path

# ---------------------------
# 1) ZIP LOCATION (update if needed)
# ---------------------------
ZIP_IN_DRIVE = "/content/drive/MyDrive/archive.zip"  # <-- update this path if your zip is elsewhere

if not os.path.exists(ZIP_IN_DRIVE):
    raise FileNotFoundError(f"ZIP not found at: {ZIP_IN_DRIVE}. Update ZIP_IN_DRIVE before running.")

# ---------------------------
# 2) Unzip to working folder
# ---------------------------
EXTRACT_DIR = "/content/Extracted_Kather"
if os.path.exists(EXTRACT_DIR):
    shutil.rmtree(EXTRACT_DIR)
os.makedirs(EXTRACT_DIR, exist_ok=True)

print(f"ğŸ“¦ Extracting {ZIP_IN_DRIVE} â†’ {EXTRACT_DIR} ...")
with zipfile.ZipFile(ZIP_IN_DRIVE, 'r') as zf:
    zf.extractall(EXTRACT_DIR)
print("âœ… Extraction complete")

# Helper: find dataset root (folder with most subfolders)
def find_dataset_root(base):
    best, best_count = base, 0
    for root, dirs, _ in os.walk(base):
        if len(dirs) > best_count:
            best, best_count = root, len(dirs)
    return best

SOURCE_DIR = find_dataset_root(EXTRACT_DIR)
print(f"âœ… Dataset root detected at: {SOURCE_DIR}")

# ==========================================================
# PARAMETERS
# ==========================================================
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

VAL_RATIO = 0.20
TEST_RATIO = 0.15
IMG_SIZE = (224, 224)
BATCH = 32
EXTS = (".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff")

# ==========================================================
# DISCOVER CLASSES & FILES
# ==========================================================
def list_images_recursive(folder):
    out = []
    for root, _, files in os.walk(folder):
        for fn in files:
            if os.path.splitext(fn)[1].lower() in EXTS:
                out.append(os.path.join(root, fn))
    return sorted(out)

class_to_files = {}
for cls in sorted(os.listdir(SOURCE_DIR)):
    p = os.path.join(SOURCE_DIR, cls)
    if os.path.isdir(p):
        files = list_images_recursive(p)
        if files:
            class_to_files[cls] = files

available_classes = sorted(class_to_files.keys())
print("\nğŸ“š Classes detected:")
for c in available_classes:
    print(f"  - {c}: {len(class_to_files[c])} images")

if len(available_classes) == 0:
    raise RuntimeError("No classes found in extracted dataset. Check the zip structure.")

# ==========================================================
# STRATIFIED SPLIT (train/val/test)
# ==========================================================
SPLIT_ROOT = "/content/kather_split_densenet"
TRAIN_DIR, VAL_DIR, TEST_DIR = [f"{SPLIT_ROOT}/{x}" for x in ["train", "val", "test"]]
if os.path.exists(SPLIT_ROOT):
    shutil.rmtree(SPLIT_ROOT)
os.makedirs(TRAIN_DIR); os.makedirs(VAL_DIR); os.makedirs(TEST_DIR)

def allocate_counts(n, val_ratio=VAL_RATIO, test_ratio=TEST_RATIO):
    if n <= 0: return (0,0,0)
    if n == 1: return (1,0,0)
    if n == 2: return (1,0,1)
    test_n = max(1, int(n * test_ratio))
    val_n  = max(1, int(n * val_ratio))
    if test_n + val_n >= n:
        val_n = 1; test_n = 1
    train_n = n - val_n - test_n
    return (train_n, val_n, test_n)

rng = np.random.default_rng(SEED)
totals = {"train":0,"val":0,"test":0}

for cls in available_classes:
    files = class_to_files[cls]
    idx = np.arange(len(files))
    rng.shuffle(idx)

    train_n, val_n, test_n = allocate_counts(len(files))
    train_idx = set(idx[:train_n])
    val_idx   = set(idx[train_n:train_n+val_n])
    test_idx  = set(idx[train_n+val_n:train_n+val_n+test_n])

    os.makedirs(f"{TRAIN_DIR}/{cls}")
    os.makedirs(f"{VAL_DIR}/{cls}")
    os.makedirs(f"{TEST_DIR}/{cls}")

    for i in idx:
        src = files[i]; base = os.path.basename(src)
        if i in train_idx: dst = f"{TRAIN_DIR}/{cls}/{base}"
        elif i in val_idx: dst = f"{VAL_DIR}/{cls}/{base}"
        else:              dst = f"{TEST_DIR}/{cls}/{base}"
        shutil.copy2(src, dst)

    totals["train"] += train_n
    totals["val"]   += val_n
    totals["test"]  += test_n

print("\nâœ… Stratified Split Completed:", totals)

# ==========================================================
# DATA GENERATORS
# ==========================================================
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=25, width_shift_range=0.2, height_shift_range=0.2,
    zoom_range=0.3, horizontal_flip=True, vertical_flip=True,
    brightness_range=[0.8,1.2], shear_range=0.2, fill_mode="reflect"
)
val_test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_directory(TRAIN_DIR, target_size=IMG_SIZE, batch_size=BATCH, class_mode='categorical', shuffle=True)
val_gen   = val_test_datagen.flow_from_directory(VAL_DIR, target_size=IMG_SIZE, batch_size=BATCH, class_mode='categorical', shuffle=False)
test_gen  = val_test_datagen.flow_from_directory(TEST_DIR, target_size=IMG_SIZE, batch_size=BATCH, class_mode='categorical', shuffle=False)
class_names = list(train_gen.class_indices.keys())
num_classes = len(class_names)
print(f"\nâœ… Detected {num_classes} classes: {class_names}")

# ==========================================================
# BUILD DENSENET MODEL
# ==========================================================
base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=IMG_SIZE + (3,))
base_model.trainable = False
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(512, activation='relu', kernel_regularizer=regularizers.l2(1e-4)),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])
model.compile(optimizer=optimizers.Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
print(model.summary())

# ==========================================================
# DQN Agent (for LR adaptation)
# ==========================================================
LR_FACTORS = np.array([0.5, 1.0, 2.0], dtype=np.float32)
MIN_LR, MAX_LR = 1e-6, 1e-3

class ReplayBuffer:
    def __init__(self, capacity=500):
        self.capacity = capacity
        self.buffer = []
        self.pos = 0
    def push(self, s, a, r, s2, done):
        if len(self.buffer) < self.capacity:
            self.buffer.append((s, a, r, s2, done))
        else:
            self.buffer[self.pos] = (s, a, r, s2, done)
            self.pos = (self.pos + 1) % self.capacity
    def sample(self, batch_size):
        idx = np.random.choice(len(self.buffer), size=batch_size, replace=False)
        s, a, r, s2, done = zip(*(self.buffer[i] for i in idx))
        return np.vstack(s), np.array(a), np.array(r, dtype=np.float32), np.vstack(s2), np.array(done, dtype=np.float32)
    def __len__(self): return len(self.buffer)

class DQNAgent:
    def __init__(self, state_dim=3, n_actions=3, lr=1e-3, gamma=0.99,
                 buffer_capacity=500, batch_size=32, epsilon_start=1.0, epsilon_final=0.05, epsilon_decay=300):
        self.n_actions = n_actions
        self.state_dim = state_dim
        self.gamma = gamma
        self.batch_size = batch_size

        inputs = Input(shape=(state_dim,))
        x = Dense(64, activation='relu')(inputs)
        x = Dense(64, activation='relu')(x)
        outputs = Dense(n_actions, activation=None)(x)
        self.q_net = Model(inputs, outputs)
        self.q_net.compile(optimizer=optimizers.Adam(learning_rate=lr), loss='mse')

        inputs_t = Input(shape=(state_dim,))
        xt = Dense(64, activation='relu')(inputs_t)
        xt = Dense(64, activation='relu')(xt)
        outputs_t = Dense(n_actions, activation=None)(xt)
        self.target_net = Model(inputs_t, outputs_t)
        self.target_net.set_weights(self.q_net.get_weights())

        self.replay = ReplayBuffer(capacity=buffer_capacity)
        self.epsilon = epsilon_start
        self.epsilon_final = epsilon_final
        self.epsilon_decay = epsilon_decay
        self.step_count = 0
        self.target_update_freq = 10

    def select_action(self, state):
        if np.random.rand() < self.epsilon:
            action = np.random.randint(self.n_actions)
        else:
            qvals = self.q_net.predict(state, verbose=0).ravel()
            action = int(np.argmax(qvals))
        qvals = self.q_net.predict(state, verbose=0).ravel()
        probs = qvals - qvals.min()
        if probs.sum() <= 0:
            probs = np.ones_like(probs) / len(probs)
        else:
            probs = probs / probs.sum()
        return action, probs

    def store_transition(self, s, a, r, s2, done=False):
        self.replay.push(s, a, r, s2, done)

    def replay_train(self, train_steps=1):
        losses = []
        for _ in range(train_steps):
            if len(self.replay) < max(4, self.batch_size):
                break
            s, a, r, s2, done = self.replay.sample(self.batch_size)
            q_s = self.q_net.predict(s, verbose=0)
            q_s2_target = self.target_net.predict(s2, verbose=0)
            target_q = q_s.copy()
            for i in range(len(a)):
                target = r[i] + self.gamma * (1.0 - done[i]) * np.max(q_s2_target[i])
                target_q[i, a[i]] = target
            history = self.q_net.train_on_batch(s, target_q)
            losses.append(float(history))
            self.step_count += 1
            self.epsilon = max(self.epsilon_final,
                               self.epsilon - (1.0 - self.epsilon_final) / self.epsilon_decay)
            if (self.step_count % self.target_update_freq) == 0:
                self.target_net.set_weights(self.q_net.get_weights())
        avg_loss = float(np.mean(losses)) if losses else 0.0
        return avg_loss

def make_state(val_acc, prev_val_acc, lr):
    delta = val_acc - prev_val_acc
    lmin, lmax = np.log10(MIN_LR), np.log10(MAX_LR)
    lrn = (np.log10(lr) - lmin) / (lmax - lmin + 1e-8)
    s = np.array([[val_acc, delta, lrn]], dtype=np.float32)
    return s

def apply_lr_action(cur_lr, action_idx):
    new_lr = float(np.clip(cur_lr * float(LR_FACTORS[action_idx]), MIN_LR, MAX_LR))
    return new_lr

agent = DQNAgent(state_dim=3, n_actions=len(LR_FACTORS), lr=1e-3,
                 buffer_capacity=500, batch_size=32, epsilon_start=1.0, epsilon_final=0.05, epsilon_decay=300)

# ==========================================================
# TRAINING (Phase 1 + Phase 2)
# ==========================================================
current_lr = float(tf.keras.backend.get_value(model.optimizer.learning_rate))
prev_val_acc = 0.0
prev_agent_state = None
prev_agent_action = None
lr_track, reward_track = [], []

# Phase 1: base frozen
EPOCHS_INITIAL = 15
for epoch in range(EPOCHS_INITIAL):
    hist = model.fit(train_gen, validation_data=val_gen, epochs=1, verbose=1)
    val_acc = float(hist.history['val_accuracy'][-1])
    current_state = make_state(val_acc, prev_val_acc, current_lr)

    if prev_agent_state is not None:
        agent.store_transition(prev_agent_state, prev_agent_action, val_acc, current_state, done=False)
        loss = agent.replay_train(train_steps=1)
    else:
        loss = 0.0

    action, _ = agent.select_action(current_state)
    new_lr = apply_lr_action(current_lr, action)
    model.optimizer.learning_rate.assign(new_lr)
    current_lr = new_lr
    prev_agent_state = current_state
    prev_agent_action = action
    prev_val_acc = val_acc
    lr_track.append(current_lr)
    reward_track.append(val_acc)
    print(f"ğŸ” Epoch {epoch+1}/{EPOCHS_INITIAL} | Val Acc: {val_acc:.4f} | New LR (DQN): {current_lr:.6f} | ReplayLoss: {loss:.6f}")

# Terminal small replay
if prev_agent_state is not None:
    agent.store_transition(prev_agent_state, prev_agent_action, val_acc, prev_agent_state, done=True)
    _ = agent.replay_train(train_steps=5)

# Phase 2: fine-tune (freeze first 100 base layers)
base_model.trainable = True
for layer in base_model.layers[:100]:
    layer.trainable = False

model.compile(optimizer=optimizers.Adam(learning_rate=current_lr), loss='categorical_crossentropy', metrics=['accuracy'])
EPOCHS_FINE = 25
for epoch in range(EPOCHS_FINE):
    hist = model.fit(train_gen, validation_data=val_gen, epochs=1, verbose=1)
    val_acc = float(hist.history['val_accuracy'][-1])
    current_state = make_state(val_acc, prev_val_acc, current_lr)

    if prev_agent_state is not None:
        agent.store_transition(prev_agent_state, prev_agent_action, val_acc, current_state, done=False)
        loss = agent.replay_train(train_steps=1)
    else:
        loss = 0.0

    action, _ = agent.select_action(current_state)
    new_lr = apply_lr_action(current_lr, action)
    model.optimizer.learning_rate.assign(new_lr)
    current_lr = new_lr
    prev_agent_state = current_state
    prev_agent_action = action
    prev_val_acc = val_acc
    lr_track.append(current_lr)
    reward_track.append(val_acc)
    print(f"ğŸ§  Fine Epoch {epoch+1}/{EPOCHS_FINE} | Val Acc: {val_acc:.4f} | New LR (DQN): {current_lr:.6f} | ReplayLoss: {loss:.6f}")

# final replay passes
if prev_agent_state is not None:
    agent.store_transition(prev_agent_state, prev_agent_action, val_acc, prev_agent_state, done=True)
    _ = agent.replay_train(train_steps=10)

# ==========================================================
# EVALUATION + PLOTS + SAVE HISTORIES
# ==========================================================
OUT_DIR = "/content/training_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

test_loss, test_acc = model.evaluate(test_gen, verbose=1)
print(f"\nâœ… Test Accuracy: {test_acc*100:.2f}%")
print(f"âœ… Test Loss: {test_loss:.4f}")

y_true = test_gen.classes
y_pred = np.argmax(model.predict(test_gen, verbose=0), axis=1)
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names,
            annot_kws={"size":10})
plt.title("Confusion Matrix â€” DenseNet121 + DQN-LR-Adapt", fontsize=16)
plt.xlabel("Predicted", fontsize=14)
plt.ylabel("Actual", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
cm_path = os.path.join(OUT_DIR, "confusion_matrix.png")
plt.savefig(cm_path, dpi=300)
print(f"ğŸ’¾ Saved confusion matrix to: {cm_path}")
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=class_names))

plt.figure(figsize=(8,5))
plt.plot(reward_track, marker='o')
plt.title("Validation Accuracy per Epoch", fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Validation Accuracy", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
acc_path = os.path.join(OUT_DIR, "val_accuracy_per_epoch.png")
plt.savefig(acc_path, dpi=300)
print(f"ğŸ’¾ Saved validation accuracy plot to: {acc_path}")
plt.show()

plt.figure(figsize=(8,5))
plt.plot(lr_track, marker='o')
plt.title("Learning Rate per Epoch (DQN)", fontsize=14)
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Learning Rate", fontsize=12)
plt.yscale('log')
plt.grid(True, which='both', linestyle='--', alpha=0.4)
plt.tight_layout()
lr_path = os.path.join(OUT_DIR, "lr_per_epoch.png")
plt.savefig(lr_path, dpi=300)
print(f"ğŸ’¾ Saved learning rate plot to: {lr_path}")
plt.show()

plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(reward_track, marker='o'); plt.title("Validation Accuracy per Epoch"); plt.grid(True)
plt.subplot(1,2,2)
plt.plot(lr_track, marker='o'); plt.yscale('log'); plt.title("Learning Rate per Epoch (DQN)"); plt.grid(True)
plt.tight_layout()
combined_path = os.path.join(OUT_DIR, "accuracy_and_lr.png")
plt.savefig(combined_path, dpi=300)
print(f"ğŸ’¾ Saved combined plot to: {combined_path}")
plt.show()

hist_df = pd.DataFrame({"val_accuracy": reward_track, "learning_rate": lr_track})
hist_csv = os.path.join(OUT_DIR, "training_history.csv")
hist_df.to_csv(hist_csv, index_label="epoch")
print(f"ğŸ’¾ Saved training history CSV to: {hist_csv}")

# ==========================================================
# SAVE MODELS: DenseNet + DQN networks to requested locations
# ==========================================================
DENSENET_DIR = "/content/models/densenet"
DQN_DIR = "/content/models/dqn_agent"
os.makedirs(DENSENET_DIR, exist_ok=True)
os.makedirs(DQN_DIR, exist_ok=True)

densenet_path = os.path.join(DENSENET_DIR, "model.h5")
model.save(densenet_path)
print(f"ğŸ’¾ Saved DenseNet model â†’ {densenet_path}")

# Save Q-net and target-net weights/models
q_net_path = os.path.join(DQN_DIR, "q_network.h5")
target_net_path = os.path.join(DQN_DIR, "target_network.h5")
agent.q_net.save(q_net_path)
agent.target_net.save(target_net_path)
print(f"ğŸ’¾ Saved DQN Q-network â†’ {q_net_path}")
print(f"ğŸ’¾ Saved DQN target-network â†’ {target_net_path}")

print("\nğŸ‰ All done. Outputs saved to:")
print(f" - Training outputs: {OUT_DIR}")
print(f" - DenseNet model: {densenet_path}")
print(f" - DQN agent: {DQN_DIR}")
```
