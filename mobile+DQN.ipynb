```python


!pip install -q tensorflow keras opencv-python matplotlib seaborn scikit-learn


from google.colab import drive
drive.mount('/content/drive')


import os, shutil, random, math, collections
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras import regularizers, optimizers

print("âœ… Libraries and Drive ready")


ZIP_IN_DRIVE = "/content/drive/MyDrive/archive.zip"  # Path to your CRC ZIP dataset
if not os.path.exists(ZIP_IN_DRIVE):
    raise FileNotFoundError("âŒ ZIP file not found in Google Drive path.")

EXTRACT_DIR = "/content/Extracted_Kather_MobileNetV2_DQN"
if os.path.exists(EXTRACT_DIR):
    shutil.rmtree(EXTRACT_DIR)
os.makedirs(EXTRACT_DIR, exist_ok=True)

print(f"ğŸ“¦ Extracting dataset from: {ZIP_IN_DRIVE}")
!unzip -q -o "$ZIP_IN_DRIVE" -d "$EXTRACT_DIR"


def find_dataset_root(base):
    best, best_count = base, 0
    for root, dirs, _ in os.walk(base):
        if len(dirs) > best_count:
            best, best_count = root, len(dirs)
    return best

SOURCE_DIR = find_dataset_root(EXTRACT_DIR)
print(f"âœ… Dataset root found at: {SOURCE_DIR}")


SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)
VAL_RATIO = 0.2
TEST_RATIO = 0.15
IMG_SIZE = (224, 224)   # MobileNetV2 default
BATCH_SIZE = 32
EXTS = (".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff")


def list_images_recursive(folder):
    out = []
    for root, _, files in os.walk(folder):
        for fn in files:
            if os.path.splitext(fn)[1].lower() in EXTS:
                out.append(os.path.join(root, fn))
    return sorted(out)

class_to_files = {}
for cls in sorted(os.listdir(SOURCE_DIR)):
    p = os.path.join(SOURCE_DIR, cls)
    if os.path.isdir(p):
        files = list_images_recursive(p)
        if files:
            class_to_files[cls] = files

available_classes = sorted(class_to_files.keys())
print("\nğŸ“š Classes Detected:")
for c in available_classes:
    print(f"  - {c}: {len(class_to_files[c])} images")

SPLIT_ROOT = "/content/kather_split_mobilenetv2_dqn"
TRAIN_DIR, VAL_DIR, TEST_DIR = [f"{SPLIT_ROOT}/{x}" for x in ["train", "val", "test"]]
if os.path.exists(SPLIT_ROOT):
    shutil.rmtree(SPLIT_ROOT)
os.makedirs(TRAIN_DIR); os.makedirs(VAL_DIR); os.makedirs(TEST_DIR)

def allocate_counts(n, val_ratio=VAL_RATIO, test_ratio=TEST_RATIO):
    if n <= 0: return (0,0,0)
    if n == 1: return (1,0,0)
    if n == 2: return (1,0,1)
    test_n = max(1, int(n * test_ratio))
    val_n  = max(1, int(n * val_ratio))
    if test_n + val_n >= n:
        val_n = 1; test_n = 1
    train_n = n - val_n - test_n
    return (train_n, val_n, test_n)

rng = np.random.default_rng(SEED)
totals = {"train":0,"val":0,"test":0}

for cls in available_classes:
    files = class_to_files[cls]
    idx = np.arange(len(files)); rng.shuffle(idx)
    train_n, val_n, test_n = allocate_counts(len(files))
    train_idx = set(idx[:train_n])
    val_idx   = set(idx[train_n:train_n+val_n])
    test_idx  = set(idx[train_n+val_n:train_n+val_n+test_n])

    os.makedirs(f"{TRAIN_DIR}/{cls}")
    os.makedirs(f"{VAL_DIR}/{cls}")
    os.makedirs(f"{TEST_DIR}/{cls}")

    for i in idx:
        src = files[i]; base = os.path.basename(src)
        if i in train_idx: dst = f"{TRAIN_DIR}/{cls}/{base}"
        elif i in val_idx: dst = f"{VAL_DIR}/{cls}/{base}"
        else:              dst = f"{TEST_DIR}/{cls}/{base}"
        shutil.copy2(src, dst)

    totals["train"] += train_n; totals["val"] += val_n; totals["test"] += test_n

print("\nâœ… Stratified Split Completed:", totals)


train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=20, width_shift_range=0.2, height_shift_range=0.2,
    zoom_range=0.2, horizontal_flip=True, vertical_flip=True,
    brightness_range=[0.8,1.2], fill_mode="nearest"
)
val_test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_directory(
    TRAIN_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', shuffle=True)
val_gen = val_test_datagen.flow_from_directory(
    VAL_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', shuffle=False)
test_gen = val_test_datagen.flow_from_directory(
    TEST_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode='categorical', shuffle=False)

class_names = list(train_gen.class_indices.keys())
num_classes = len(class_names)
print(f"\nâœ… Number of Classes: {num_classes}\n")


base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=IMG_SIZE + (3,))
base_model.trainable = False  # Freeze for initial training

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(256, activation='relu', kernel_regularizer=regularizers.l2(1e-4)),
    Dropout(0.4),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

INIT_LR = 1e-4
model.compile(optimizer=optimizers.Adam(learning_rate=INIT_LR),
              loss='categorical_crossentropy', metrics=['accuracy'])

print(model.summary())


class ReplayBuffer:
    def __init__(self, capacity=1024):
        self.buf = collections.deque(maxlen=capacity)
    def push(self, s, a, r, s2, d):
        self.buf.append((s.astype(np.float32), np.int32(a), np.float32(r), s2.astype(np.float32), np.float32(d)))
    def sample(self, batch):
        n = min(batch, len(self.buf))
        idx = np.random.choice(len(self.buf), size=n, replace=False)
        s,a,r,s2,d = zip(*[self.buf[i] for i in idx])
        return np.stack(s), np.array(a), np.array(r)[:,None], np.stack(s2), np.array(d)[:,None]
    def __len__(self): return len(self.buf)

def q_net(hidden=[128,128], n_actions=5):
    i = tf.keras.Input(shape=(3,))  # [val_acc, best_val_acc, log_lr]
    x = i
    for u in hidden:
        x = tf.keras.layers.Dense(u, activation='relu')(x)
    o = tf.keras.layers.Dense(n_actions)(x)
    return tf.keras.Model(i, o)

class DQNAgent:
    def __init__(self, lr_bounds=(1e-6, 3e-3), init_lr=1e-4, gamma=0.98, tau=0.01,
                 epsilon_start=1.0, epsilon_end=0.05, epsilon_decay=0.90,
                 n_actions=5, q_lr=1e-3):
        # Actions in log-space: multiply LR by these factors
        self.actions_log = np.log(np.array([0.5, 0.8, 1.0, 1.2, 1.5], dtype=np.float32))
        assert n_actions == len(self.actions_log)
        self.n_actions = n_actions

        self.lr_min, self.lr_max = lr_bounds
        self.log_lr = float(np.log(init_lr))

        self.q = q_net([128,128], n_actions)
        self.q_t = q_net([128,128], n_actions)
        self.q_t.set_weights(self.q.get_weights())
        self.opt = tf.keras.optimizers.Adam(q_lr)

        self.gamma = gamma
        self.tau = tau

        self.epsilon = epsilon_start
        self.epsilon_end = epsilon_end
        self.epsilon_decay = epsilon_decay

        self.rb = ReplayBuffer(2048)

    def select_action(self, s):
        # Îµ-greedy on Q(s,Â·)
        if np.random.rand() < self.epsilon:
            a = np.random.randint(self.n_actions)
        else:
            qvals = self.q(np.asarray(s, dtype=np.float32)[None,...]).numpy()[0]
            a = int(np.argmax(qvals))
        return a

    def step(self, val_acc, best_val_acc):
        s = np.array([val_acc, best_val_acc, self.log_lr], dtype=np.float32)
        a = self.select_action(s)
        new_log_lr = self.log_lr + float(self.actions_log[a])
        new_lr = float(np.clip(np.exp(new_log_lr), self.lr_min, self.lr_max))
        self.log_lr = float(np.log(new_lr))  # reflect clipping
        return new_lr, a, s

    def update(self, batch_size=32, updates=4):
        if len(self.rb) < 16:
            return
        for _ in range(updates):
            s, a, r, s2, d = self.rb.sample(batch_size)
            with tf.GradientTape() as tape:
                q_s = self.q(s)                          # (B, A)
                q_s_a = tf.gather(q_s, a, batch_dims=1)  # (B,)

                # target: r + Î³*(1-d)*max_a' Q_t(s', a')
                q_s2_t = self.q_t(s2)
                max_q_s2 = tf.reduce_max(q_s2_t, axis=1, keepdims=True)
                y = r + (1.0 - d) * self.gamma * max_q_s2

                loss = tf.reduce_mean((q_s_a[:,None] - tf.stop_gradient(y))**2)

            grads = tape.gradient(loss, self.q.trainable_variables)
            self.opt.apply_gradients(zip(grads, self.q.trainable_variables))

            # Soft-update target
            for wt, w in zip(self.q_t.weights, self.q.weights):
                wt.assign(self.tau * w + (1.0 - self.tau) * wt)

        # decay epsilon
        self.epsilon = max(self.epsilon_end, self.epsilon * self.epsilon_decay)

dqn = DQNAgent(lr_bounds=(1e-6, 3e-3), init_lr=INIT_LR,
               epsilon_start=1.0, epsilon_end=0.05, epsilon_decay=0.90)

# ==========================================================
# ğŸ‹ï¸ Stage 1 â€” Train (Feature Extraction) with DQN LR
# (epochs unchanged: 8)
# ==========================================================
EPOCHS_INITIAL = 15
best_val_acc = 0.0
lr_track, reward_track = [], []

for epoch in range(EPOCHS_INITIAL):
    hist = model.fit(train_gen, validation_data=val_gen, epochs=1, verbose=1)
    val_acc = float(hist.history['val_accuracy'][-1])
    reward = val_acc - best_val_acc  # improvement signal
    done = 1.0 if reward <= 0 else 0.0

    # DQN step -> new LR
    new_lr, action_idx, state = dqn.step(val_acc, best_val_acc)
    next_state = np.array([val_acc, max(best_val_acc, val_acc), dqn.log_lr], dtype=np.float32)

    # Store transition and update Q
    dqn.rb.push(state, action_idx, reward, next_state, done)
    dqn.update(batch_size=32, updates=4)

    model.optimizer.learning_rate.assign(new_lr)
    lr_track.append(new_lr); reward_track.append(reward)

    if val_acc > best_val_acc:
        best_val_acc = val_acc

    print(f"ğŸ® Epoch {epoch+1}/{EPOCHS_INITIAL} | val_acc={val_acc:.4f} | Î”best={reward:+.4f} | LRâ†’ {new_lr:.2e} | eps={dqn.epsilon:.2f}")

# ==========================================================
# ğŸ”“ Stage 2 â€” Fine-tuning (Top Layers) with DQN LR
# (epochs unchanged: 15)
# ==========================================================
base_model.trainable = True
for layer in base_model.layers[:100]:  # freeze early layers (unchanged)
    layer.trainable = False

model.compile(optimizer=optimizers.Adam(learning_rate=np.exp(dqn.log_lr)),
              loss='categorical_crossentropy', metrics=['accuracy'])

EPOCHS_FINE = 25
for epoch in range(EPOCHS_FINE):
    hist = model.fit(train_gen, validation_data=val_gen, epochs=1, verbose=1)
    val_acc = float(hist.history['val_accuracy'][-1])
    reward = val_acc - best_val_acc
    done = 1.0 if reward <= 0 else 0.0

    new_lr, action_idx, state = dqn.step(val_acc, best_val_acc)
    next_state = np.array([val_acc, max(best_val_acc, val_acc), dqn.log_lr], dtype=np.float32)

    dqn.rb.push(state, action_idx, reward, next_state, done)
    dqn.update(batch_size=32, updates=6)

    model.optimizer.learning_rate.assign(new_lr)
    lr_track.append(new_lr); reward_track.append(reward)

    if val_acc > best_val_acc:
        best_val_acc = val_acc

    print(f"ğŸ§  Fine {epoch+1}/{EPOCHS_FINE} | val_acc={val_acc:.4f} | Î”best={reward:+.4f} | LRâ†’ {new_lr:.2e} | eps={dqn.epsilon:.2f}")

# ==========================================================
# ğŸ§¾ Evaluation
# ==========================================================
test_loss, test_acc = model.evaluate(test_gen, verbose=1)
print(f"\nâœ… Test Accuracy: {test_acc*100:.2f}%")
print(f"âœ… Test Loss: {test_loss:.4f}")

y_true = test_gen.classes
y_pred = np.argmax(model.predict(test_gen, verbose=0), axis=1)
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix â€” MobileNetV2 + DQN")
plt.xlabel("Predicted"); plt.ylabel("Actual")
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=class_names))

# ==========================================================
# ğŸ“Š DQN Dynamics
# ==========================================================
plt.figure(figsize=(12,4))
plt.plot(reward_track, label='Reward (Î” val_acc)')
plt.title("DQN Reward per Epoch"); plt.legend(); plt.show()

plt.figure(figsize=(12,4))
plt.plot(lr_track, label='Learning Rate (DQN)')
plt.title("DQN Adaptive Learning Rate"); plt.legend(); plt.show()

# ==========================================================
# ğŸ’¾ Save Model
# ==========================================================
model.save('/content/mobilenetv2_crc_DQN_finetuned.keras')
print("âœ… Model saved at: /content/mobilenetv2_crc_DQN_finetuned.keras")
```
