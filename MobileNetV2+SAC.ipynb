```python
# Single-cell Colab-ready script: MobileNetV2 + Full-featured SAC (discrete LR multipliers)


# -------------------- Colab Drive mount (robust) --------------------
MOUNT_POINT = '/content/drive'
try:
    from google.colab import drive
    try:
        drive.mount(MOUNT_POINT, force_remount=True)
    except Exception as e:
        import os, shutil
        try:
            if os.path.exists(MOUNT_POINT):
                try:
                    shutil.rmtree(MOUNT_POINT)
                except Exception:
                    pass
            os.makedirs(MOUNT_POINT, exist_ok=True)
            drive.mount(MOUNT_POINT, force_remount=True)
        except Exception as e2:
            raise RuntimeError(f"Drive mount failed: {e} / {e2}")
except Exception as e:
    print("Warning: google.colab.drive not available. Are you running this in Colab?", e)

# -------------------- imports --------------------
import os, glob, shutil, random, zipfile, math, json, datetime, itertools, sys, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, optimizers, regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import tensorflow_probability as tfp

print("âœ… Libraries ready, TF version:", tf.__version__)

# -------------------- USER EDIT --------------------
ZIP_IN_DRIVE = "/content/drive/MyDrive/archive.zip"   # <<-- edit to your zip
PROJECT_DIR  = "/content/drive/MyDrive/RL_CNN_Research/mobilenetv2_sac_enhanced"  # top-level project folder
os.makedirs(PROJECT_DIR, exist_ok=True)
RESULTS_MASTER = os.path.join(PROJECT_DIR, "results_master.csv")
# ---------------------------------------------------

# -------------------- per-run folder --------------------
run_ts = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
rand_suf = f"{np.random.randint(1000,9999)}"
RUN_ID = f"run_{run_ts}_{rand_suf}"
RUN_DIR = os.path.join(PROJECT_DIR, RUN_ID)
os.makedirs(RUN_DIR, exist_ok=True)

# redirect commonly used paths into RUN_DIR
CKPT_DIR = os.path.join(RUN_DIR, "checkpoints"); os.makedirs(CKPT_DIR, exist_ok=True)
LOG_CSV = os.path.join(RUN_DIR, "sac_enhanced_log.csv")
BEST_MODEL_PATH = os.path.join(RUN_DIR, "mobilenetv2_sac_enhanced_best.keras")
FINAL_MODEL_PATH = os.path.join(RUN_DIR, "mobilenetv2_sac_enhanced_final.keras")
PER_CLASS_CSV = os.path.join(RUN_DIR, "per_class_report.csv")
MISCLASSIFIED_DIR = os.path.join(RUN_DIR, "misclassified"); os.makedirs(MISCLASSIFIED_DIR, exist_ok=True)

# record run info
with open(os.path.join(RUN_DIR, "run_info.txt"), "w") as f:
    f.write(f"RUN_ID: {RUN_ID}\nPROJECT_DIR: {PROJECT_DIR}\nZIP_IN_DRIVE: {ZIP_IN_DRIVE}\nTIMESTAMP: {run_ts}\n")

# -------------------- config & seeds --------------------
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)
VAL_RATIO = 0.20; TEST_RATIO = 0.15
IMG_SIZE = (224,224); BATCH_SIZE = 32
EXTS = (".jpg",".jpeg",".png",".bmp",".tif",".tiff")

# -------------------- unzip dataset --------------------
EXTRACT_DIR = "/content/Extracted_Mobile_SAC_Enhanced"
if os.path.exists(EXTRACT_DIR): shutil.rmtree(EXTRACT_DIR)
os.makedirs(EXTRACT_DIR, exist_ok=True)

if not os.path.exists(ZIP_IN_DRIVE):
    raise FileNotFoundError(f"ZIP not found at {ZIP_IN_DRIVE}. Did you set ZIP_IN_DRIVE correctly and mount Drive?")
print("Unzipping... (this may take a while for big zips)")
with zipfile.ZipFile(ZIP_IN_DRIVE, 'r') as zf:
    zf.extractall(EXTRACT_DIR)
print("Unzipped.")

def list_images_recursive(folder):
    out=[]
    for r,_,files in os.walk(folder):
        for f in files:
            if os.path.splitext(f)[1].lower() in EXTS:
                out.append(os.path.join(r,f))
    return sorted(out)

def find_dataset_root(base):
    candidates=[]
    for root, dirs, _ in os.walk(base):
        if not dirs: continue
        score=0
        for d in dirs:
            dpath=os.path.join(root,d)
            for f in os.listdir(dpath):
                if os.path.splitext(f)[1].lower() in EXTS:
                    score+=1; break
        if score>0: candidates.append((root,score))
    if candidates:
        candidates.sort(key=lambda x:x[1], reverse=True); return candidates[0][0]
    return base

DATA_ROOT = find_dataset_root(EXTRACT_DIR)
print("DATA_ROOT:", DATA_ROOT)

# -------------------- map classes & split --------------------
class_to_files={}
for entry in sorted(os.listdir(DATA_ROOT)):
    p=os.path.join(DATA_ROOT, entry)
    if os.path.isdir(p):
        imgs=list_images_recursive(p)
        if imgs: class_to_files[entry]=imgs
if not class_to_files:
    # fallback scan deeper
    for r,dirs,files in os.walk(DATA_ROOT):
        for d in dirs:
            p=os.path.join(r,d)
            imgs=list_images_recursive(p)
            if imgs: class_to_files[d]=imgs

available_classes = sorted(class_to_files.keys())
print("Classes and counts:", [(c,len(class_to_files[c])) for c in available_classes])

SPLIT_ROOT = "/content/mobilenet_sac_enhanced_split"
if os.path.exists(SPLIT_ROOT): shutil.rmtree(SPLIT_ROOT)
TRAIN_DIR = os.path.join(SPLIT_ROOT,"train"); VAL_DIR = os.path.join(SPLIT_ROOT,"val"); TEST_DIR = os.path.join(SPLIT_ROOT,"test")
os.makedirs(TRAIN_DIR, exist_ok=True); os.makedirs(VAL_DIR, exist_ok=True); os.makedirs(TEST_DIR, exist_ok=True)

def allocate_counts(n,val_ratio=VAL_RATIO,test_ratio=TEST_RATIO):
    if n<=0: return (0,0,0)
    if n==1: return (1,0,0)
    if n==2: return (1,0,1)
    test_n=max(1,int(round(n*test_ratio)))
    val_n=max(1,int(round(n*val_ratio)))
    if test_n+val_n>=n:
        # ensure at least one train
        if n>=3:
            val_n=1; test_n=1
        else:
            # fallback safe split
            val_n=0; test_n=1
    train_n=n-val_n-test_n; return (train_n,val_n,test_n)

rng=np.random.default_rng(SEED)
tot={"train":0,"val":0,"test":0}
for cls, files in class_to_files.items():
    files = sorted(files); n=len(files); idx=np.arange(n); rng.shuffle(idx)
    tr_n, v_n, te_n = allocate_counts(n)
    tr_idx=set(idx[:tr_n]); v_idx=set(idx[tr_n:tr_n+v_n]); te_idx=set(idx[tr_n+v_n:tr_n+v_n+te_n])
    os.makedirs(os.path.join(TRAIN_DIR,cls), exist_ok=True)
    os.makedirs(os.path.join(VAL_DIR,cls), exist_ok=True)
    os.makedirs(os.path.join(TEST_DIR,cls), exist_ok=True)
    for i in idx:
        src=files[i]; basename=os.path.basename(src)
        if i in tr_idx: dst=os.path.join(TRAIN_DIR,cls,basename)
        elif i in v_idx: dst=os.path.join(VAL_DIR,cls,basename)
        else: dst=os.path.join(TEST_DIR,cls,basename)
        try:
            shutil.copy2(src,dst)
        except Exception:
            # if copy fails, still continue
            pass
    tot['train']+=len(tr_idx); tot['val']+=len(v_idx); tot['test']+=len(te_idx)
print("Split done:", tot)

# -------------------- Enhanced Data Augmentation --------------------
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=25,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=False,
    brightness_range=[0.8, 1.2],
    channel_shift_range=0.2,
    fill_mode='nearest'
)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_directory(TRAIN_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=True)
val_gen   = val_datagen.flow_from_directory(VAL_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False)
test_gen  = val_datagen.flow_from_directory(TEST_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False)
class_names = list(train_gen.class_indices.keys()); num_classes=len(class_names)
print("Detected classes:", class_names)

# save class_indices mapping for reproducibility
try:
    json.dump(train_gen.class_indices, open(os.path.join(RUN_DIR,'class_indices.json'),'w'))
except Exception:
    pass

# -------------------- Enhanced MobileNetV2 classifier --------------------
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=IMG_SIZE+(3,))
base_model.trainable=False

inputs = keras.Input(shape=IMG_SIZE+(3,))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(1024, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
x = layers.Dropout(0.6)(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
x = layers.Dropout(0.5)(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation='softmax')(x)
model = keras.Model(inputs, outputs)

INITIAL_LR = 1e-4
model.compile(
    optimizer=optimizers.Adam(learning_rate=INITIAL_LR, beta_1=0.9, beta_2=0.999),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()

# -------------------- Full SAC (Soft Actor-Critic) Implementation --------------------
ACTIONS = [0.3, 0.5, 0.75, 1.0, 1.25, 1.5, 2.0]  # More granular control
N_ACTIONS = len(ACTIONS)
STATE_DIM = 4  # Added momentum for better state representation

# Enhanced SAC hyperparameters
GAMMA = 0.99
TAU = 0.005  # Soft update coefficient
INITIAL_ALPHA = 0.2  # Initial temperature parameter
TARGET_ENTROPY = -0.98 * np.log(1.0 / N_ACTIONS)  # Target entropy for automatic alpha tuning
LR_ACTOR = 1e-4
LR_CRITIC = 1e-3
LR_ALPHA = 3e-4
REPLAY_BUFFER_SIZE = 50000
SAC_BATCH_SIZE = 512  # Larger batch for stability
WARMUP_STEPS = 2000
TARGET_UPDATE_INTERVAL = 1
GRAD_CLIP_NORM = 1.0
LEARN_ALPHA = True  # Automatically tune temperature parameter

# Training phases
PHASE1_EPOCHS = 15
PHASE2_EPOCHS = 25

# -------------------- Enhanced SAC Networks --------------------
class QNetwork(keras.Model):
    def __init__(self, state_dim, action_dim, hidden_dim=512):
        super(QNetwork, self).__init__()
        self.input_layer = layers.Dense(hidden_dim, activation='relu')
        self.layer_norm1 = layers.LayerNormalization()
        self.hidden1 = layers.Dense(hidden_dim, activation='relu')
        self.layer_norm2 = layers.LayerNormalization()
        self.hidden2 = layers.Dense(hidden_dim, activation='relu')
        self.output_layer = layers.Dense(action_dim)

    def call(self, state):
        x = self.input_layer(state)
        x = self.layer_norm1(x)
        x = self.hidden1(x)
        x = self.layer_norm2(x)
        x = self.hidden2(x)
        return self.output_layer(x)

class ActorNetwork(keras.Model):
    def __init__(self, state_dim, action_dim, hidden_dim=512):
        super(ActorNetwork, self).__init__()
        self.action_dim = action_dim
        self.input_layer = layers.Dense(hidden_dim, activation='relu')
        self.layer_norm1 = layers.LayerNormalization()
        self.hidden1 = layers.Dense(hidden_dim, activation='relu')
        self.layer_norm2 = layers.LayerNormalization()
        self.hidden2 = layers.Dense(hidden_dim, activation='relu')
        self.logits_layer = layers.Dense(action_dim)

    def call(self, state):
        x = self.input_layer(state)
        x = self.layer_norm1(x)
        x = self.hidden1(x)
        x = self.layer_norm2(x)
        x = self.hidden2(x)
        logits = self.logits_layer(x)
        return tf.nn.softmax(logits, axis=-1)

    def sample_action(self, state):
        probs = self.call(state)
        action_dist = tfp.distributions.Categorical(probs=probs)
        action = action_dist.sample()
        log_prob = action_dist.log_prob(action)
        return action, log_prob, probs

# Initialize networks with larger capacity
q_network1 = QNetwork(STATE_DIM, N_ACTIONS, hidden_dim=512)
q_network2 = QNetwork(STATE_DIM, N_ACTIONS, hidden_dim=512)
target_q_network1 = QNetwork(STATE_DIM, N_ACTIONS, hidden_dim=512)
target_q_network2 = QNetwork(STATE_DIM, N_ACTIONS, hidden_dim=512)
actor_network = ActorNetwork(STATE_DIM, N_ACTIONS, hidden_dim=512)

# Initialize target networks with same weights
target_q_network1.set_weights(q_network1.get_weights())
target_q_network2.set_weights(q_network2.get_weights())

# Optimizers with weight decay
actor_optimizer = optimizers.Adam(learning_rate=LR_ACTOR)
critic_optimizer1 = optimizers.Adam(learning_rate=LR_CRITIC)
critic_optimizer2 = optimizers.Adam(learning_rate=LR_CRITIC)

# Temperature parameter (learnable)
log_alpha = tf.Variable(tf.math.log(INITIAL_ALPHA), trainable=LEARN_ALPHA)
alpha_optimizer = optimizers.Adam(learning_rate=LR_ALPHA)

# -------------------- Prioritized Experience Replay Buffer --------------------
class PrioritizedReplayBuffer:
    def __init__(self, buffer_size, state_dim, alpha=0.6, beta=0.4, beta_increment=0.001):
        self.buffer_size = buffer_size
        self.state_dim = state_dim
        self.alpha = alpha
        self.beta = beta
        self.beta_increment = beta_increment
        self.pos = 0
        self.size = 0

        self.states = np.zeros((buffer_size, state_dim), dtype=np.float32)
        self.actions = np.zeros(buffer_size, dtype=np.int32)
        self.rewards = np.zeros(buffer_size, dtype=np.float32)
        self.next_states = np.zeros((buffer_size, state_dim), dtype=np.float32)
        self.dones = np.zeros(buffer_size, dtype=np.float32)
        self.priorities = np.zeros(buffer_size, dtype=np.float32)
        self.max_priority = 1.0

    def add(self, state, action, reward, next_state, done):
        self.states[self.pos] = state
        self.actions[self.pos] = action
        self.rewards[self.pos] = reward
        self.next_states[self.pos] = next_state
        self.dones[self.pos] = done
        self.priorities[self.pos] = self.max_priority

        self.pos = (self.pos + 1) % self.buffer_size
        self.size = min(self.size + 1, self.buffer_size)

    def sample(self, batch_size):
        if self.size == 0:
            return None

        priorities = self.priorities[:self.size]
        probs = priorities ** self.alpha
        probs /= probs.sum()

        indices = np.random.choice(self.size, batch_size, p=probs)
        weights = (self.size * probs[indices]) ** (-self.beta)
        weights /= weights.max()

        self.beta = min(1.0, self.beta + self.beta_increment)

        return (
            tf.convert_to_tensor(self.states[indices]),
            tf.convert_to_tensor(self.actions[indices]),
            tf.convert_to_tensor(self.rewards[indices]),
            tf.convert_to_tensor(self.next_states[indices]),
            tf.convert_to_tensor(self.dones[indices]),
            indices,
            tf.convert_to_tensor(weights, dtype=tf.float32)
        )

    def update_priorities(self, indices, priorities):
        for idx, priority in zip(indices, priorities):
            self.priorities[idx] = priority + 1e-5  # Small constant to avoid zero priority
            self.max_priority = max(self.max_priority, priority)

# Initialize prioritized replay buffer
replay_buffer = PrioritizedReplayBuffer(REPLAY_BUFFER_SIZE, STATE_DIM)

# -------------------- Enhanced SAC Training Functions --------------------
@tf.function
def update_critic(states, actions, rewards, next_states, dones, weights):
    with tf.GradientTape(persistent=True) as tape:
        # Sample actions from current policy for next states
        next_action_probs = actor_network(next_states)
        next_action_dist = tfp.distributions.Categorical(probs=next_action_probs)
        next_actions = next_action_dist.sample()
        next_log_probs = next_action_dist.log_prob(next_actions)

        # Compute target Q values
        target_q1 = target_q_network1(next_states)
        target_q2 = target_q_network2(next_states)
        target_q = tf.minimum(target_q1, target_q2)

        target_q_selected = tf.reduce_sum(
            tf.one_hot(next_actions, N_ACTIONS) * target_q, axis=1
        )

        current_alpha = tf.exp(log_alpha)
        target_q_values = rewards + GAMMA * (1 - dones) * (
            target_q_selected - current_alpha * next_log_probs
        )

        # Current Q values
        current_q1 = q_network1(states)
        current_q2 = q_network2(states)

        current_q1_selected = tf.reduce_sum(
            tf.one_hot(actions, N_ACTIONS) * current_q1, axis=1
        )
        current_q2_selected = tf.reduce_sum(
            tf.one_hot(actions, N_ACTIONS) * current_q2, axis=1
        )

        # Critic losses with TD errors for prioritization
        td_error1 = tf.abs(current_q1_selected - target_q_values)
        td_error2 = tf.abs(current_q2_selected - target_q_values)

        critic_loss1 = tf.reduce_mean(weights * tf.square(current_q1_selected - target_q_values))
        critic_loss2 = tf.reduce_mean(weights * tf.square(current_q2_selected - target_q_values))

    # Apply gradients with clipping
    critic_grads1 = tape.gradient(critic_loss1, q_network1.trainable_variables)
    critic_grads2 = tape.gradient(critic_loss2, q_network2.trainable_variables)

    if GRAD_CLIP_NORM > 0:
        critic_grads1, _ = tf.clip_by_global_norm(critic_grads1, GRAD_CLIP_NORM)
        critic_grads2, _ = tf.clip_by_global_norm(critic_grads2, GRAD_CLIP_NORM)

    critic_optimizer1.apply_gradients(zip(critic_grads1, q_network1.trainable_variables))
    critic_optimizer2.apply_gradients(zip(critic_grads2, q_network2.trainable_variables))

    return critic_loss1 + critic_loss2, td_error1 + td_error2

@tf.function
def update_actor(states, weights):
    with tf.GradientTape() as tape:
        action_probs = actor_network(states)
        action_dist = tfp.distributions.Categorical(probs=action_probs)
        actions = action_dist.sample()
        log_probs = action_dist.log_prob(actions)

        q1 = q_network1(states)
        q2 = q_network2(states)
        q = tf.minimum(q1, q2)

        q_selected = tf.reduce_sum(tf.one_hot(actions, N_ACTIONS) * q, axis=1)
        current_alpha = tf.exp(log_alpha)

        actor_loss = tf.reduce_mean(weights * (current_alpha * log_probs - q_selected))

    actor_grads = tape.gradient(actor_loss, actor_network.trainable_variables)
    if GRAD_CLIP_NORM > 0:
        actor_grads, _ = tf.clip_by_global_norm(actor_grads, GRAD_CLIP_NORM)
    actor_optimizer.apply_gradients(zip(actor_grads, actor_network.trainable_variables))

    return actor_loss, log_probs

@tf.function
def update_alpha(log_probs):
    if LEARN_ALPHA:
        with tf.GradientTape() as tape:
            alpha_loss = -tf.reduce_mean(log_alpha * (log_probs + TARGET_ENTROPY))

        alpha_grads = tape.gradient(alpha_loss, [log_alpha])
        alpha_optimizer.apply_gradients(zip(alpha_grads, [log_alpha]))
        return alpha_loss
    return 0.0

def soft_update(target_network, source_network, tau):
    for target_param, source_param in zip(target_network.variables, source_network.variables):
        target_param.assign(tau * source_param + (1.0 - tau) * target_param)

# -------------------- Enhanced State Representation --------------------
def enhanced_state_from(val_acc, prev_val_acc, prev_val_acc2, lr, momentum=0.0):
    delta = float(val_acc - prev_val_acc)
    delta2 = float(prev_val_acc - prev_val_acc2) if prev_val_acc2 is not None else 0.0
    lmin, lmax = math.log10(1e-6), math.log10(1e-3)
    lrn = (math.log10(max(lr,1e-12)) - lmin) / (lmax - lmin + 1e-8)
    return np.array([float(val_acc), float(delta), float(delta2), float(momentum)], dtype=np.float32)

# -------------------- Enhanced Training Loop --------------------
def set_optimizer_lr(optim, lr):
    try:
        optim.learning_rate.assign(lr)
    except Exception:
        try:
            tf.keras.backend.set_value(optim.learning_rate, lr)
        except Exception:
            try: optim.learning_rate = lr
            except Exception: pass

def save_checkpoint_weights(model_obj, phase, update_idx, keep_last=3):
    fname = os.path.join(CKPT_DIR, f"{phase}_update{update_idx}.weights.h5")
    try:
        model_obj.save_weights(fname)
    except Exception as e:
        print("Warning: failed to save weights:", e)
    pattern = os.path.join(CKPT_DIR, f"{phase}_update*.weights.h5")
    all_ck = sorted([p for p in glob.glob(pattern)], key=os.path.getmtime, reverse=True)
    for old in all_ck[keep_last:]:
        try: os.remove(old)
        except: pass
    return fname

# -------------------- SAC Training Procedure --------------------
def collect_experience_and_train(use_random_policy=False, training_step=0):
    global current_lr, prev_val_acc, prev_val_acc2, best_val_acc

    # Train for one epoch
    train_steps = max(1, len(train_gen))
    val_steps = max(1, len(val_gen))

    # Add learning rate scheduler callback for additional stability
    reduce_lr = keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', factor=0.5, patience=3,
        min_lr=1e-7, verbose=1
    )

    h = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=1,
        verbose=1,
        steps_per_epoch=train_steps,
        validation_steps=val_steps,
        callbacks=[reduce_lr]
    )

    val_acc = float(h.history.get('val_accuracy', [0.0])[-1])
    val_loss = float(h.history.get('val_loss', [0.0])[-1])

    # Enhanced state with momentum
    momentum = 0.9 * (prev_val_acc - prev_val_acc2) if prev_val_acc2 is not None else 0.0
    s = enhanced_state_from(val_acc, prev_val_acc, prev_val_acc2, current_lr, momentum)

    # Choose action
    if use_random_policy or training_step < WARMUP_STEPS:
        action_idx = np.random.randint(0, N_ACTIONS)
        log_prob = math.log(1.0 / N_ACTIONS)
    else:
        state_tensor = tf.convert_to_tensor(s.reshape(1, -1), dtype=tf.float32)
        action_idx, log_prob, _ = actor_network.sample_action(state_tensor)
        action_idx = action_idx.numpy()[0]
        log_prob = log_prob.numpy()[0]

    action_multiplier = ACTIONS[action_idx]
    new_lr = float(np.clip(current_lr * action_multiplier, 1e-7, 1e-2))
    set_optimizer_lr(model.optimizer, new_lr)
    current_lr = new_lr

    # Enhanced reward shaping
    reward = float(val_acc)  # Base reward
    if prev_val_acc2 is not None:
        # Additional reward for consistent improvement
        improvement = val_acc - prev_val_acc
        if improvement > 0:
            reward += 0.1 * improvement  # Small bonus for improvement
        elif improvement < -0.05:  # Penalty for significant degradation
            reward += 0.5 * improvement

    # Next state (using current values for next state)
    next_momentum = 0.9 * momentum + 0.1 * (val_acc - prev_val_acc)
    next_s = enhanced_state_from(val_acc, prev_val_acc, prev_val_acc2, current_lr, next_momentum)

    # Store experience
    replay_buffer.add(s, action_idx, reward, next_s, 0.0)

    # Update previous accuracies
    prev_val_acc2 = prev_val_acc
    prev_val_acc = val_acc

    # Train SAC networks if we have enough samples
    if replay_buffer.size >= SAC_BATCH_SIZE and training_step >= WARMUP_STEPS:
        for _ in range(2):  # Multiple updates per step
            batch_data = replay_buffer.sample(SAC_BATCH_SIZE)
            if batch_data is not None:
                states, actions, rewards, next_states, dones, indices, weights = batch_data

                # Update critics
                critic_loss, td_errors = update_critic(states, actions, rewards, next_states, dones, weights)

                # Update actor and alpha
                actor_loss, log_probs = update_actor(states, weights)
                alpha_loss = update_alpha(log_probs)

                # Update priorities
                replay_buffer.update_priorities(indices, td_errors.numpy())

                # Soft update target networks
                if training_step % TARGET_UPDATE_INTERVAL == 0:
                    soft_update(target_q_network1, q_network1, TAU)
                    soft_update(target_q_network2, q_network2, TAU)

    # Update best validation accuracy
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        try:
            model.save(BEST_MODEL_PATH)
            print(f"ğŸ¯ New best model! val_acc={val_acc:.4f}")
        except Exception as e:
            print("Warning: failed to save best model:", e)

    return {
        'val_acc': val_acc,
        'val_loss': val_loss,
        'action': action_idx,
        'reward': reward,
        'lr': current_lr,
        'alpha': float(tf.exp(log_alpha).numpy())
    }

# -------------------- Main Training Loop --------------------
# Initialize log CSV
if not os.path.exists(LOG_CSV):
    pd.DataFrame(columns=[
        'timestamp', 'phase', 'step', 'val_acc', 'val_loss', 'reward',
        'action', 'lr', 'alpha', 'replay_size', 'best_val_acc'
    ]).to_csv(LOG_CSV, index=False)

# Training state
best_val_acc = 0.0
current_lr = INITIAL_LR
prev_val_acc = 0.0
prev_val_acc2 = None
training_step = 0

def save_final_and_append_results(test_acc, mis_csv):
    try:
        model.save(FINAL_MODEL_PATH)
    except Exception as e:
        print("Warning: failed to save final model:", e)
    final_row = {
        'timestamp': datetime.datetime.now().isoformat(),
        'method': 'MobileNetV2+SAC_Enhanced',
        'test_accuracy': float(test_acc),
        'num_classes': int(num_classes),
        'best_val_accuracy': float(best_val_acc),
        'final_model_path': FINAL_MODEL_PATH,
        'best_model_path': BEST_MODEL_PATH,
        'log_csv': LOG_CSV,
        'per_class_report': PER_CLASS_CSV,
        'misclassified_dir': MISCLASSIFIED_DIR,
        'misclassified_list_csv': mis_csv,
        'run_dir': RUN_DIR
    }
    try:
        if not os.path.exists(RESULTS_MASTER):
            pd.DataFrame([final_row]).to_csv(RESULTS_MASTER, index=False)
        else:
            dfm = pd.read_csv(RESULTS_MASTER)
            dfm = pd.concat([dfm, pd.DataFrame([final_row])], ignore_index=True)
            dfm.to_csv(RESULTS_MASTER, index=False)
        print("Appended run summary to:", RESULTS_MASTER)
    except Exception as e:
        print("Warning: couldn't append to results_master:", e)

def train_phase_with_sac(phase_name, total_epochs):
    global training_step, prev_val_acc, prev_val_acc2

    print(f"\nğŸ¯ Starting {phase_name} phase with SAC for {total_epochs} epochs")

    for epoch in range(total_epochs):
        print(f"\n=== {phase_name} Epoch {epoch+1}/{total_epochs} ===")

        use_random = training_step < WARMUP_STEPS
        result = collect_experience_and_train(use_random_policy=use_random, training_step=training_step)

        # Log results
        log_row = {
            'timestamp': str(datetime.datetime.now()),
            'phase': phase_name,
            'step': training_step,
            'val_acc': float(result['val_acc']),
            'val_loss': float(result['val_loss']),
            'reward': float(result['reward']),
            'action': int(result['action']),
            'lr': float(result['lr']),
            'alpha': float(result['alpha']),
            'replay_size': int(replay_buffer.size),
            'best_val_acc': float(best_val_acc)
        }

        df = pd.read_csv(LOG_CSV)
        df = pd.concat([df, pd.DataFrame([log_row])], ignore_index=True)
        df.to_csv(LOG_CSV, index=False)

        print(f"Epoch {epoch+1}: val_acc={result['val_acc']:.4f}, val_loss={result['val_loss']:.4f}, "
              f"lr={result['lr']:.6f}, action={ACTIONS[result['action']]}, "
              f"alpha={result['alpha']:.4f}, replay_size={replay_buffer.size}")

        training_step += 1

        # Save checkpoint every 5 epochs
        if (epoch + 1) % 5 == 0:
            save_checkpoint_weights(model, phase_name, epoch + 1)

print("=== Phase 1: Frozen Training with SAC ===")
train_phase_with_sac("frozen", PHASE1_EPOCHS)

print("=== Phase 2: Fine-tuning with SAC ===")
# Unfreeze more layers for fine-tuning
base_model.trainable = True
for layer in base_model.layers[:-30]:  # Unfreeze more layers
    layer.trainable = False
for layer in base_model.layers[-30:]:
    layer.trainable = True


model.compile(
    optimizer=optimizers.Adam(learning_rate=current_lr, beta_1=0.9, beta_2=0.999),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

train_phase_with_sac("finetune", PHASE2_EPOCHS)

print("=== Final Evaluation ===")

if len(test_gen) == 0:
    print("No test images found. Skipping evaluation.")
    test_acc = 0.0
    y_true = np.array([])
    y_pred = np.array([])
    mis_rows = []
    mis_csv = ""
else:

    try:
        best_model = keras.models.load_model(BEST_MODEL_PATH)
        test_res = best_model.evaluate(test_gen, verbose=1, steps=max(1, len(test_gen)))
    except:
        test_res = model.evaluate(test_gen, verbose=1, steps=max(1, len(test_gen)))
        best_model = model

    test_acc = float(test_res[1]) if len(test_res) > 1 else float(test_res[0])

    y_true = np.array(test_gen.classes)
    y_pred_proba = best_model.predict(test_gen, verbose=0, steps=max(1, len(test_gen)))
    y_pred = np.argmax(y_pred_proba, axis=1)

    try:
        report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
        report_df = pd.DataFrame(report).transpose()
        report_df.to_csv(PER_CLASS_CSV)

        accuracy_per_class = []
        for i, class_name in enumerate(class_names):
            class_mask = y_true == i
            if np.sum(class_mask) > 0:
                class_accuracy = np.mean(y_pred[class_mask] == i)
                accuracy_per_class.append(class_accuracy)
            else:
                accuracy_per_class.append(0.0)

        print(f"\nğŸ“Š Final Test Accuracy: {test_acc:.4f}")
        print(f"ğŸ“ˆ Best Validation Accuracy: {best_val_acc:.4f}")
        print(f"ğŸ¯ Class-wise Accuracy: {np.mean(accuracy_per_class):.4f} Â± {np.std(accuracy_per_class):.4f}")

    except Exception as e:
        print("Warning: failed to save classification report:", e)

    mis_rows = []
    filenames = test_gen.filepaths
    confidence_scores = np.max(y_pred_proba, axis=1)

    for i, fp in enumerate(filenames):
        if i >= len(y_pred): break
        t = int(y_true[i]); p = int(y_pred[i])
        if p != t:
            mis_rows.append({
                'filepath': fp,
                'true': class_names[t],
                'pred': class_names[p],
                'confidence': float(confidence_scores[i]),
                'confidence_true': float(y_pred_proba[i, t])
            })
            dst = os.path.join(MISCLASSIFIED_DIR,
                             f"true_{class_names[t]}_pred_{class_names[p]}_conf_{confidence_scores[i]:.3f}_{os.path.basename(fp)}")
            try:
                shutil.copy2(fp, dst)
            except Exception:
                pass

    mis_csv = os.path.join(RUN_DIR, "misclassified_analysis.csv")
    try:
        pd.DataFrame(mis_rows).to_csv(mis_csv, index=False)
        print(f"ğŸ“ Saved {len(mis_rows)} misclassified examples analysis")
    except Exception:
        mis_csv = ""


    try:
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(12, 10))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=class_names, yticklabels=class_names,
                   cbar_kws={'label': 'Count'})
        plt.title(f"Confusion Matrix â€” MobileNetV2 + Enhanced SAC\nFinal Test Accuracy: {test_acc:.4f}", fontsize=14)
        plt.xlabel("Predicted", fontsize=12)
        plt.ylabel("Actual", fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.savefig(os.path.join(RUN_DIR, "confusion_matrix.png"), dpi=300, bbox_inches='tight')
        plt.show()
    except Exception as e:
        print("Warning: couldn't plot confusion matrix:", e)

    try:
        log_df = pd.read_csv(LOG_CSV)
        plt.figure(figsize=(15, 10))

        plt.subplot(2, 2, 1)
        plt.plot(log_df['val_acc'], label='Validation Accuracy')
        plt.plot(log_df['best_val_acc'], label='Best Validation Accuracy', linestyle='--')
        plt.axhline(y=test_acc, color='r', linestyle='-', label=f'Final Test Accuracy: {test_acc:.4f}')
        plt.xlabel('Training Step')
        plt.ylabel('Accuracy')
        plt.title('Training Progress')
        plt.legend()
        plt.grid(True)

        plt.subplot(2, 2, 2)
        plt.plot(log_df['val_loss'])
        plt.xlabel('Training Step')
        plt.ylabel('Validation Loss')
        plt.title('Validation Loss')
        plt.grid(True)

        plt.subplot(2, 2, 3)
        plt.plot(log_df['lr'])
        plt.xlabel('Training Step')
        plt.ylabel('Learning Rate')
        plt.title('Learning Rate Schedule')
        plt.yscale('log')
        plt.grid(True)

        plt.subplot(2, 2, 4)
        plt.plot(log_df['reward'])
        plt.xlabel('Training Step')
        plt.ylabel('Reward')
        plt.title('SAC Rewards')
        plt.grid(True)

        plt.tight_layout()
        plt.savefig(os.path.join(RUN_DIR, "training_history.png"), dpi=300, bbox_inches='tight')
        plt.show()
    except Exception as e:
        print("Warning: couldn't plot training history:", e)

save_final_and_append_results(test_acc, mis_csv if len(mis_rows) > 0 else "")

print(f"\nğŸ‰ Training Complete!")
print(f"ğŸ“Š Final Test Accuracy: {test_acc:.4f}")
print(f"ğŸ† Best Validation Accuracy: {best_val_acc:.4f}")
print(f"ğŸ’¾ Results saved in: {RUN_DIR}")
print(f"ğŸ“ˆ Log file: {LOG_CSV}")
print(f"ğŸ”§ Best model: {BEST_MODEL_PATH}")
```

```python
# -------------------- Colab Drive mount (robust) --------------------
MOUNT_POINT = '/content/drive'
try:
    from google.colab import drive
    try:
        drive.mount(MOUNT_POINT, force_remount=True)
    except Exception as e:
        import os, shutil
        try:
            if os.path.exists(MOUNT_POINT):
                try:
                    shutil.rmtree(MOUNT_POINT)
                except Exception:
                    pass
            os.makedirs(MOUNT_POINT, exist_ok=True)
            drive.mount(MOUNT_POINT, force_remount=True)
        except Exception as e2:
            raise RuntimeError(f"Drive mount failed: {e} / {e2}")
except Exception as e:
    print("Warning: google.colab.drive not available. Are you running this in Colab?", e)

# -------------------- imports --------------------
import os, glob, shutil, random, zipfile, math, json, datetime, itertools, sys, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, optimizers, regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

print("âœ… Libraries ready, TF version:", tf.__version__)

# -------------------- USER EDIT --------------------
ZIP_IN_DRIVE = "/content/drive/MyDrive/archive.zip"   # <<-- edit to your zip
PROJECT_DIR  = "/content/drive/MyDrive/RL_CNN_Research/mobilenetv2_sac_allfeatures"  # top-level project folder
os.makedirs(PROJECT_DIR, exist_ok=True)
RESULTS_MASTER = os.path.join(PROJECT_DIR, "results_master.csv")
# ---------------------------------------------------

# -------------------- per-run folder --------------------
run_ts = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
rand_suf = f"{np.random.randint(1000,9999)}"
RUN_ID = f"run_{run_ts}_{rand_suf}"
RUN_DIR = os.path.join(PROJECT_DIR, RUN_ID)
os.makedirs(RUN_DIR, exist_ok=True)

CKPT_DIR = os.path.join(RUN_DIR, "checkpoints"); os.makedirs(CKPT_DIR, exist_ok=True)
LOG_CSV = os.path.join(RUN_DIR, "sac_allfeatures_log.csv")
BEST_MODEL_PATH = os.path.join(RUN_DIR, "mobilenetv2_sac_allfeatures_best.keras")
FINAL_MODEL_PATH = os.path.join(RUN_DIR, "mobilenetv2_sac_allfeatures_final.keras")
PER_CLASS_CSV = os.path.join(RUN_DIR, "per_class_report.csv")
MISCLASSIFIED_DIR = os.path.join(RUN_DIR, "misclassified"); os.makedirs(MISCLASSIFIED_DIR, exist_ok=True)

with open(os.path.join(RUN_DIR, "run_info.txt"), "w") as f:
    f.write(f"RUN_ID: {RUN_ID}\nPROJECT_DIR: {PROJECT_DIR}\nZIP_IN_DRIVE: {ZIP_IN_DRIVE}\nTIMESTAMP: {run_ts}\n")

# -------------------- config & seeds --------------------
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)
VAL_RATIO = 0.20; TEST_RATIO = 0.15
IMG_SIZE = (224,224); BATCH_SIZE = 32
EXTS = (".jpg",".jpeg",".png",".bmp",".tif",".tiff")

# -------------------- unzip dataset --------------------
EXTRACT_DIR = "/content/Extracted_Mobile_SAC_AllFeatures"
if os.path.exists(EXTRACT_DIR): shutil.rmtree(EXTRACT_DIR)
os.makedirs(EXTRACT_DIR, exist_ok=True)

if not os.path.exists(ZIP_IN_DRIVE):
    raise FileNotFoundError(f"ZIP not found at {ZIP_IN_DRIVE}. Did you set ZIP_IN_DRIVE correctly and mount Drive?")
print("Unzipping... (this may take a while for big zips)")
with zipfile.ZipFile(ZIP_IN_DRIVE, 'r') as zf:
    zf.extractall(EXTRACT_DIR)
print("Unzipped.")

def list_images_recursive(folder):
    out=[]
    for r,_,files in os.walk(folder):
        for f in files:
            if os.path.splitext(f)[1].lower() in EXTS:
                out.append(os.path.join(r,f))
    return sorted(out)

def find_dataset_root(base):
    candidates=[]
    for root, dirs, _ in os.walk(base):
        if not dirs: continue
        score=0
        for d in dirs:
            dpath=os.path.join(root,d)
            for f in os.listdir(dpath):
                if os.path.splitext(f)[1].lower() in EXTS:
                    score+=1; break
        if score>0: candidates.append((root,score))
    if candidates:
        candidates.sort(key=lambda x:x[1], reverse=True); return candidates[0][0]
    return base

DATA_ROOT = find_dataset_root(EXTRACT_DIR)
print("DATA_ROOT:", DATA_ROOT)

# -------------------- map classes & split --------------------
class_to_files={}
for entry in sorted(os.listdir(DATA_ROOT)):
    p=os.path.join(DATA_ROOT, entry)
    if os.path.isdir(p):
        imgs=list_images_recursive(p)
        if imgs: class_to_files[entry]=imgs
if not class_to_files:
    for r,dirs,files in os.walk(DATA_ROOT):
        for d in dirs:
            p=os.path.join(r,d)
            imgs=list_images_recursive(p)
            if imgs: class_to_files[d]=imgs

available_classes = sorted(class_to_files.keys())
print("Classes and counts:", [(c,len(class_to_files[c])) for c in available_classes])

SPLIT_ROOT = "/content/mobilenet_sac_allfeatures_split"
if os.path.exists(SPLIT_ROOT): shutil.rmtree(SPLIT_ROOT)
TRAIN_DIR = os.path.join(SPLIT_ROOT,"train"); VAL_DIR = os.path.join(SPLIT_ROOT,"val"); TEST_DIR = os.path.join(SPLIT_ROOT,"test")
os.makedirs(TRAIN_DIR, exist_ok=True); os.makedirs(VAL_DIR, exist_ok=True); os.makedirs(TEST_DIR, exist_ok=True)

def allocate_counts(n,val_ratio=VAL_RATIO,test_ratio=TEST_RATIO):
    if n<=0: return (0,0,0)
    if n==1: return (1,0,0)
    if n==2: return (1,0,1)
    test_n=max(1,int(round(n*test_ratio)))
    val_n=max(1,int(round(n*val_ratio)))
    if test_n+val_n>=n:
        if n>=3:
            val_n=1; test_n=1
        else:
            val_n=0; test_n=1
    train_n=n-val_n-test_n; return (train_n,val_n,test_n)

rng=np.random.default_rng(SEED)
tot={"train":0,"val":0,"test":0}
for cls, files in class_to_files.items():
    files = sorted(files); n=len(files); idx=np.arange(n); rng.shuffle(idx)
    tr_n, v_n, te_n = allocate_counts(n)
    tr_idx=set(idx[:tr_n]); v_idx=set(idx[tr_n:tr_n+v_n]); te_idx=set(idx[tr_n+v_n:tr_n+v_n+te_n])
    os.makedirs(os.path.join(TRAIN_DIR,cls), exist_ok=True)
    os.makedirs(os.path.join(VAL_DIR,cls), exist_ok=True)
    os.makedirs(os.path.join(TEST_DIR,cls), exist_ok=True)
    for i in idx:
        src=files[i]; basename=os.path.basename(src)
        if i in tr_idx: dst=os.path.join(TRAIN_DIR,cls,basename)
        elif i in v_idx: dst=os.path.join(VAL_DIR,cls,basename)
        else: dst=os.path.join(TEST_DIR,cls,basename)
        try:
            shutil.copy2(src,dst)
        except Exception:
            pass
    tot['train']+=len(tr_idx); tot['val']+=len(v_idx); tot['test']+=len(te_idx)
print("Split done:", tot)

# -------------------- datagens & strong augmentation --------------------
# keep same base generator but we'll wrap with MixUp/CutMix
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input,
                                   rotation_range=20, width_shift_range=0.1, height_shift_range=0.1,
                                   zoom_range=0.15, horizontal_flip=True)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_directory(TRAIN_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=True)
val_gen   = val_datagen.flow_from_directory(VAL_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False)
test_gen  = val_datagen.flow_from_directory(TEST_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False)
class_names = list(train_gen.class_indices.keys()); num_classes=len(class_names)
print("Detected classes:", class_names)

try:
    json.dump(train_gen.class_indices, open(os.path.join(RUN_DIR,'class_indices.json'),'w'))
except Exception:
    pass

# MixUp + CutMix utilities
def mixup_cutmix_batch(x, y, alpha=0.4, prob_cutmix=0.5):
    N = x.shape[0]
    lam = np.random.beta(alpha, alpha) if alpha>0 else 1.0
    idx = np.random.permutation(N)
    x2 = x[idx]; y2 = y[idx]
    if np.random.rand() < prob_cutmix:
        H, W = x.shape[1], x.shape[2]
        cut_rat = np.sqrt(1. - lam)
        cut_w = int(W * cut_rat)
        cut_h = int(H * cut_rat)
        cx = np.random.randint(W)
        cy = np.random.randint(H)
        x1 = np.clip(cx - cut_w // 2, 0, W)
        x2c = np.clip(cx + cut_w // 2, 0, W)
        y1 = np.clip(cy - cut_h // 2, 0, H)
        y2c = np.clip(cy + cut_h // 2, 0, H)
        x_new = x.copy()
        x_new[:, y1:y2c, x1:x2c, :] = x2[:, y1:y2c, x1:x2c, :]
        lam_adj = 1.0 - ((x2c - x1) * (y2c - y1) / float(W * H))
        y_new = lam_adj * y + (1.0 - lam_adj) * y2
        return x_new, y_new
    else:
        x_new = lam * x + (1 - lam) * x2
        y_new = lam * y + (1 - lam) * y2
        return x_new, y_new

def mixed_generator(gen, alpha=0.4, prob_cutmix=0.5):
    while True:
        x_batch, y_batch = next(gen)
        x_m, y_m = mixup_cutmix_batch(x_batch, y_batch, alpha=alpha, prob_cutmix=prob_cutmix)
        yield x_m, y_m

train_gen_mixed = mixed_generator(train_gen, alpha=0.4, prob_cutmix=0.5)

# -------------------- build MobileNetV2 classifier (same head) --------------------
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=IMG_SIZE+(3,))
base_model.trainable=False
inputs = keras.Input(shape=IMG_SIZE+(3,))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation='softmax')(x)
model = keras.Model(inputs, outputs)

# label smoothing loss
from tensorflow.keras.losses import CategoricalCrossentropy
LABEL_SMOOTH = 0.08
loss_fn = CategoricalCrossentropy(label_smoothing=LABEL_SMOOTH)

INITIAL_LR = 1e-4
model.compile(optimizer=optimizers.Adam(learning_rate=INITIAL_LR), loss=loss_fn, metrics=['accuracy'])
model.summary()

# -------------------- SAC hyperparams & advanced features --------------------
ACTION_LOW = 0.5
ACTION_HIGH = 1.5
ACTION_DIM = 1
STATE_DIM = 3

# Keep same rollout/phase epochs to be directly comparable
ROLLOUT_STEPS = 6        # epochs per environment step (kept same as your PPO)
PHASE1_EPOCHS = 15
PHASE2_EPOCHS = 25

# SAC core params
REPLAY_SIZE = 20000
BATCH_SIZE_SAC = 64
N_STEP = 3
START_STEPS = 6
UPDATES_PER_STEP = 16
GAMMA = 0.99
POLYAK = 0.995
POLICY_LR = 3e-4
Q_LR = 3e-4
ALPHA_LR = 1e-4
TARGET_ENTROPY = -float(ACTION_DIM)

# Prioritized Replay params
PER_ALPHA = 0.6
PER_BETA_START = 0.4
PER_BETA_FRAMES = 20000
PER_EPS = 1e-6

MAX_GRAD_NORM = 0.5
LOG_STD_MIN = -20
LOG_STD_MAX = 2

# SAM option (disabled by default because requires special training loop)
USE_SAM = False

# -------------------- actor & critic networks --------------------
def build_actor(state_dim, action_dim):
    s = keras.Input(shape=(state_dim,))
    x = layers.Dense(256, activation='relu')(s)
    x = layers.Dense(256, activation='relu')(x)
    mu = layers.Dense(action_dim)(x)
    log_std = layers.Dense(action_dim)(x)
    return keras.Model(s, [mu, log_std])

def build_q_network(state_dim, action_dim):
    s = keras.Input(shape=(state_dim,))
    a = keras.Input(shape=(action_dim,))
    x = layers.Concatenate()([s,a])
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dense(256, activation='relu')(x)
    q = layers.Dense(1)(x)
    return keras.Model([s,a], q)

actor = build_actor(STATE_DIM, ACTION_DIM)
q1 = build_q_network(STATE_DIM, ACTION_DIM)
q2 = build_q_network(STATE_DIM, ACTION_DIM)
q1_target = build_q_network(STATE_DIM, ACTION_DIM)
q2_target = build_q_network(STATE_DIM, ACTION_DIM)
q1_target.set_weights(q1.get_weights())
q2_target.set_weights(q2.get_weights())

actor_opt = optimizers.Adam(learning_rate=POLICY_LR)
q1_opt = optimizers.Adam(learning_rate=Q_LR)
q2_opt = optimizers.Adam(learning_rate=Q_LR)
alpha_opt = optimizers.Adam(learning_rate=ALPHA_LR)
log_alpha = tf.Variable(0.0, dtype=tf.float32)  # automatic entropy

# -------------------- Running mean/std --------------------
class RunningMeanStd:
    def __init__(self, shape=(), eps=1e-4):
        self.mean = np.zeros(shape, 'float64')
        self.var = np.ones(shape, 'float64')
        self.count = eps
    def update(self, x):
        x = np.array(x, dtype=np.float64)
        if x.ndim == 0:
            x = x.reshape(1)
        batch_mean = np.mean(x, axis=0)
        batch_var = np.var(x, axis=0)
        batch_count = x.shape[0]
        self._update_from_moments(batch_mean, batch_var, batch_count)
    def _update_from_moments(self, batch_mean, batch_var, batch_count):
        delta = batch_mean - self.mean
        tot_count = self.count + batch_count
        new_mean = self.mean + delta * batch_count / tot_count
        m_a = self.var * (self.count)
        m_b = batch_var * (batch_count)
        M2 = m_a + m_b + np.square(delta) * (self.count * batch_count / tot_count)
        new_var = M2 / (tot_count)
        self.mean = new_mean
        self.var = new_var
        self.count = tot_count
    def normalize(self, x):
        x = np.array(x, dtype=np.float64)
        return (x - self.mean) / (np.sqrt(self.var) + 1e-8)

obs_rms = RunningMeanStd(shape=(STATE_DIM,))
rew_rms = RunningMeanStd(shape=())

# -------------------- Prioritized Replay (proportional) + n-step --------------------
class PrioritizedReplayBuffer:
    def __init__(self, size=REPLAY_SIZE, alpha=PER_ALPHA, n_step=N_STEP, gamma=GAMMA):
        self.size = size
        self.alpha = alpha
        self.n_step = n_step
        self.gamma = gamma
        self.pos = 0
        self.curr_size = 0
        self.max_priority = 1.0
        self.sum_tree = np.zeros(2*size, dtype=np.float32)  # simple over-alloc (not exact seg tree for simplicity)
        # storage
        self.states = np.zeros((size, STATE_DIM), dtype=np.float32)
        self.actions = np.zeros((size, ACTION_DIM), dtype=np.float32)
        self.rewards = np.zeros((size,), dtype=np.float32)
        self.next_states = np.zeros((size, STATE_DIM), dtype=np.float32)
        self.dones = np.zeros((size,), dtype=np.float32)
        # n-step buffer
        self.nstep_buffer = []

    def add(self, s, a, r, s2, d):
        # store n-step temporarily
        self.nstep_buffer.append((s,a,r,s2,d))
        if len(self.nstep_buffer) < self.n_step: # wait until buffer full
            return
        # compute n-step return
        R = 0.0
        for idx, (_,_,rr,_,_) in enumerate(self.nstep_buffer):
            R += (self.gamma**idx) * rr
        s0, a0, _, _, _ = self.nstep_buffer[0]
        sN, _, _, sN2, dN = self.nstep_buffer[-1]
        self.states[self.pos] = s0
        self.actions[self.pos] = a0
        self.rewards[self.pos] = R
        self.next_states[self.pos] = sN2
        self.dones[self.pos] = float(dN)
        self.max_priority = max(self.max_priority, 1.0)
        # naive priority store: sum_tree index mapping simplified
        tree_idx = self.pos + self.size
        self.sum_tree[tree_idx] = self.max_priority ** self.alpha
        # update parents
        idx2 = tree_idx // 2
        while idx2 >= 1:
            self.sum_tree[idx2] = self.sum_tree[2*idx2] + self.sum_tree[2*idx2+1]
            idx2 //= 2
        self.pos = (self.pos + 1) % self.size
        self.curr_size = min(self.curr_size + 1, self.size)
        # pop oldest n-step
        self.nstep_buffer.pop(0)

    def _sample_proportional(self, batch_size):
        # sample proportionally to priorities by binary search on sum tree (simplified)
        res_idx = []
        total = self.sum_tree[1]
        for _ in range(batch_size):
            mass = random.random() * total
            idx = 1
            while idx < self.size:
                left = 2*idx
                if self.sum_tree[left] >= mass:
                    idx = left
                else:
                    mass -= self.sum_tree[left]
                    idx = left+1
            data_idx = idx - self.size
            res_idx.append(int(data_idx))
        return res_idx

    def sample(self, batch_size, beta):
        assert self.curr_size > 0
        idxs = self._sample_proportional(batch_size)
        states = self.states[idxs]
        actions = self.actions[idxs]
        rewards = self.rewards[idxs]
        next_states = self.next_states[idxs]
        dones = self.dones[idxs]
        # importance-sampling weights
        probs = self.sum_tree[np.array(idxs)+self.size] / (self.sum_tree[1] + 1e-12)
        weights = (self.curr_size * probs) ** (-beta)
        weights /= weights.max()
        return dict(states=states, actions=actions, rewards=rewards, next_states=next_states, dones=dones, idxs=idxs, weights=weights)

    def update_priorities(self, idxs, priorities):
        for i,p in zip(idxs, priorities):
            tree_idx = int(i) + self.size
            self.sum_tree[tree_idx] = (abs(p) + PER_EPS) ** self.alpha
            idx2 = tree_idx//2
            while idx2 >= 1:
                self.sum_tree[idx2] = self.sum_tree[2*idx2] + self.sum_tree[2*idx2+1]
                idx2 //= 2
            self.max_priority = max(self.max_priority, p)

    def __len__(self):
        return self.curr_size

replay = PrioritizedReplayBuffer(size=REPLAY_SIZE, alpha=PER_ALPHA, n_step=N_STEP, gamma=GAMMA)

# -------------------- helpers & checkpoint saver --------------------
def set_optimizer_lr(optim, lr):
    try:
        optim.learning_rate.assign(lr)
    except Exception:
        try:
            tf.keras.backend.set_value(optim.learning_rate, lr)
        except Exception:
            try: optim.learning_rate = lr
            except Exception: pass

def state_from(val_acc, prev_val_acc_local, lr):
    delta = float(val_acc - prev_val_acc_local)
    lmin, lmax = math.log10(1e-6), math.log10(1e-3)
    lrn = (math.log10(max(lr,1e-12)) - lmin) / (lmax - lmin + 1e-8)
    return np.array([float(val_acc), float(delta), float(lrn)], dtype=np.float32)

def save_checkpoint_weights(model_obj, phase, update_idx, keep_last=5):
    fname = os.path.join(CKPT_DIR, f"{phase}_update{update_idx}.keras")
    try:
        model_obj.save(fname)
    except Exception as e:
        print("Warning: failed to save full model:", e)
        # fallback to weights
        try:
            wfname = fname + ".weights.h5"
            model_obj.save_weights(wfname)
            fname = wfname
        except Exception as e2:
            print("Warning: failed to save weights:", e2)
    files = sorted(glob.glob(os.path.join(CKPT_DIR, f"{phase}_update*")), key=os.path.getmtime, reverse=True)
    for old in files[keep_last:]:
        try: os.remove(old)
        except: pass
    return fname

# -------------------- SAC utilities (sampling, logp) --------------------
@tf.function
def gaussian_log_prob(mu, log_std, value):
    pre_sum = -0.5 * (((value - mu) / (tf.exp(log_std) + 1e-8))**2 + 2*log_std + tf.math.log(2*math.pi))
    return tf.reduce_sum(pre_sum, axis=1, keepdims=True)

@tf.function
def sample_action_tf(state):
    mu, log_std = actor(state)
    log_std = tf.clip_by_value(log_std, LOG_STD_MIN, LOG_STD_MAX)
    std = tf.exp(log_std)
    eps = tf.random.normal(tf.shape(mu))
    pre_tanh = mu + eps * std
    action = tf.tanh(pre_tanh)
    # scale to [ACTION_LOW, ACTION_HIGH]
    action_scaled = ACTION_LOW + (action + 1.0) * 0.5 * (ACTION_HIGH - ACTION_LOW)
    # log prob with tanh correction
    logp = gaussian_log_prob(mu, log_std, pre_tanh) - tf.reduce_sum(tf.math.log(1 - action**2 + 1e-6), axis=1, keepdims=True)
    return action_scaled, logp, mu, log_std

# -------------------- SAC update step --------------------
@tf.function
def sac_update_step(states, actions, rewards, next_states, dones, weights, alpha):
    states_t = tf.convert_to_tensor(states, dtype=tf.float32)
    actions_t = tf.convert_to_tensor(actions, dtype=tf.float32)
    rewards_t = tf.reshape(tf.convert_to_tensor(rewards, dtype=tf.float32), (-1,1))
    next_states_t = tf.convert_to_tensor(next_states, dtype=tf.float32)
    dones_t = tf.reshape(tf.convert_to_tensor(dones, dtype=tf.float32), (-1,1))
    weights_t = tf.reshape(tf.convert_to_tensor(weights, dtype=tf.float32), (-1,1))

    # target actions & logp
    next_action, next_logp, _, _ = sample_action_tf(next_states_t)
    q1_t = q1_target([next_states_t, next_action])
    q2_t = q2_target([next_states_t, next_action])
    q_target = tf.minimum(q1_t, q2_t) - alpha * next_logp
    target_q = rewards_t + (1.0 - dones_t) * (GAMMA ** N_STEP) * q_target

    # update Q networks
    with tf.GradientTape() as tape1, tf.GradientTape() as tape2:
        q1_pred = q1([states_t, actions_t])
        q2_pred = q2([states_t, actions_t])
        td1 = q1_pred - target_q
        td2 = q2_pred - target_q
        q1_loss = tf.reduce_mean(weights_t * 0.5 * tf.square(td1))
        q2_loss = tf.reduce_mean(weights_t * 0.5 * tf.square(td2))
    grads1 = tape1.gradient(q1_loss, q1.trainable_variables)
    grads2 = tape2.gradient(q2_loss, q2.trainable_variables)
    grads1, _ = tf.clip_by_global_norm(grads1, MAX_GRAD_NORM)
    grads2, _ = tf.clip_by_global_norm(grads2, MAX_GRAD_NORM)
    q1_opt.apply_gradients(zip(grads1, q1.trainable_variables))
    q2_opt.apply_gradients(zip(grads2, q2.trainable_variables))

    # policy update
    with tf.GradientTape() as tape_pi:
        pi_action, logp_pi, _, _ = sample_action_tf(states_t)
        q1_pi = q1([states_t, pi_action])
        q2_pi = q2([states_t, pi_action])
        q_pi = tf.minimum(q1_pi, q2_pi)
        policy_loss = tf.reduce_mean(weights_t * (alpha * logp_pi - q_pi))
    pi_grads = tape_pi.gradient(policy_loss, actor.trainable_variables)
    pi_grads, _ = tf.clip_by_global_norm(pi_grads, MAX_GRAD_NORM)
    actor_opt.apply_gradients(zip(pi_grads, actor.trainable_variables))

    # entropy alpha loss
    with tf.GradientTape() as tape_a:
        _, logp_samp, _, _ = sample_action_tf(states_t)
        alpha_loss = -tf.reduce_mean(log_alpha * (logp_samp + TARGET_ENTROPY))
    grads_a = tape_a.gradient(alpha_loss, [log_alpha])
    alpha_opt.apply_gradients(zip(grads_a, [log_alpha]))

    return q1_loss, q2_loss, policy_loss, tf.reduce_mean(logp_pi)

# -------------------- training loop using SAC controller --------------------
if not os.path.exists(LOG_CSV):
    pd.DataFrame(columns=[
        'timestamp','phase','update_idx','mean_val_acc','lr','chosen_actions','avg_reward',
        'q1_loss','q2_loss','policy_loss','logp','alpha'
    ]).to_csv(LOG_CSV,index=False)

best_val_acc = -1.0
current_lr = INITIAL_LR
prev_val_acc = 0.0
update_idx = 0
global_step = 0

# PER beta annealing helper
def get_per_beta(frame_idx):
    frac = min(1.0, frame_idx / float(PER_BETA_FRAMES))
    return PER_BETA_START + frac * (1.0 - PER_BETA_START)

def train_sac_phase(phase_name, total_epochs, allow_random_warmup=True):
    global prev_val_acc, best_val_acc, current_lr, update_idx, global_step
    epochs_done = 0
    while epochs_done < total_epochs:
        print(f"\n=== {phase_name} update {update_idx+1} ===")
        # collect ROLLOUT_STEPS epochs (each epoch = 1 fit on train_gen)
        for step in range(ROLLOUT_STEPS):
            # train classifier for one epoch but with MixUp generator
            train_steps = max(1, len(train_gen))
            val_steps = max(1, len(val_gen))
            # use train_gen_mixed (mixup) to improve augmentation
            h = model.fit(train_gen_mixed, validation_data=val_gen, epochs=1, verbose=1,
                          steps_per_epoch=train_steps, validation_steps=val_steps)
            val_acc = float(h.history.get('val_accuracy', [0.0])[-1])
            val_loss = float(h.history.get('val_loss', [0.0])[-1])

            # build state and normalize
            s = state_from(val_acc, prev_val_acc, current_lr)
            obs_rms.update(s.reshape(1,-1))
            s_norm = obs_rms.normalize(s)
            s_tf = s_norm.reshape(1,-1).astype(np.float32)

            # choose action (random warmup or policy)
            if allow_random_warmup and update_idx < START_STEPS:
                action = np.random.uniform(ACTION_LOW, ACTION_HIGH, size=(1,ACTION_DIM)).astype(np.float32)
                logp = math.log(1.0/(ACTION_HIGH-ACTION_LOW))
            else:
                a_samp, logp_tf, _, _ = sample_action_tf(tf.convert_to_tensor(s_tf))
                action = a_samp.numpy()
                logp = float(logp_tf.numpy())

            # apply LR multiplier
            new_lr = float(np.clip(current_lr * float(action.flatten()[0]), 1e-6, 1e-3))
            set_optimizer_lr(model.optimizer, new_lr)
            current_lr = new_lr

            # reward (use val_acc improvement + smoothing)
            reward = float(val_acc)
            rew_rms.update(np.array([reward]))
            reward_norm = float(rew_rms.normalize(np.array([reward]))[0])

            # next state (we use immediate next as placeholder)
            next_s = state_from(val_acc, val_acc, current_lr)
            next_s = obs_rms.normalize(next_s)

            done = 0.0

            # store transition (replay handles n-step internally)
            replay.add(s_norm.astype(np.float32), action.reshape(ACTION_DIM).astype(np.float32), reward_norm, next_s.astype(np.float32), done)

            prev_val_acc = val_acc
            epochs_done += 1
            global_step += 1

            # perform many updates for sample-efficiency if buffer large enough
            if len(replay) >= max(BATCH_SIZE_SAC, 100):
                per_beta = get_per_beta(global_step)
                for _ in range(UPDATES_PER_STEP):
                    batch = replay.sample(BATCH_SIZE_SAC, beta=per_beta)
                    q1_l, q2_l, p_l, mlogp = sac_update_step(batch['states'], batch['actions'], batch['rewards'], batch['next_states'], batch['dones'], batch['weights'], float(tf.exp(log_alpha).numpy()))
                    # update priorities using TD errors (sum of abs td1+td2)
                    # for simplicity use q1_l and q2_l as proxy; in practice compute per-sample td errors
                    # dummy priorities update here (could store per-sample errors)
                    # replay.update_priorities(batch['idxs'], np.ones_like(batch['idxs'])*0.5)

                # soft-update targets
                for t_var, var in zip(q1_target.trainable_variables, q1.trainable_variables):
                    t_var.assign(POLYAK * t_var + (1-POLYAK) * var)
                for t_var, var in zip(q2_target.trainable_variables, q2.trainable_variables):
                    t_var.assign(POLYAK * t_var + (1-POLYAK) * var)

        # End of rollout steps: logging & checkpointing
        avg_reward = float(np.mean(replay.rewards[:replay.curr_size])) if replay.curr_size>0 else 0.0
        q1_l_val = float(q1_l.numpy()) if 'q1_l' in locals() else 0.0
        q2_l_val = float(q2_l.numpy()) if 'q2_l' in locals() else 0.0
        p_l_val = float(p_l.numpy()) if 'p_l' in locals() else 0.0
        mlogp_val = float(mlogp.numpy()) if 'mlogp' in locals() else 0.0
        alpha_val = float(tf.exp(log_alpha).numpy())

        save_checkpoint_weights(model, phase_name, update_idx, keep_last=5)

        if prev_val_acc > best_val_acc:
            best_val_acc = prev_val_acc
            try:
                model.save(BEST_MODEL_PATH)
                print(f"Saved new best model (val_acc={best_val_acc:.4f})")
            except Exception as e:
                print("Warning: couldn't save best model:", e)

        row = {
            'timestamp': str(datetime.datetime.now()),
            'phase': phase_name,
            'update_idx': update_idx,
            'mean_val_acc': float(prev_val_acc),
            'lr': float(current_lr),
            'chosen_actions': json.dumps([float(action.flatten()[0])]) if 'action' in locals() else "[]",
            'avg_reward': avg_reward,
            'q1_loss': q1_l_val,
            'q2_loss': q2_l_val,
            'policy_loss': p_l_val,
            'logp': mlogp_val,
            'alpha': alpha_val
        }
        df = pd.read_csv(LOG_CSV)
        df = pd.concat([df, pd.DataFrame([row])], ignore_index=True)
        df.to_csv(LOG_CSV, index=False)

        print(f"[{phase_name}] update {update_idx} | val_acc={prev_val_acc:.4f} | lr={current_lr:.6f} | action={row['chosen_actions']} | alpha={alpha_val:.6f}")
        update_idx += 1

# -------------------- Run phases (same epochs as PPO) --------------------
print("=== Phase 1: frozen training (SAC controls LR) ===")
train_sac_phase("frozen", PHASE1_EPOCHS, allow_random_warmup=True)

print("=== Phase 2: fine-tune (unfreeze top layers) ===")
base_model.trainable=True
for layer in base_model.layers[:-50]:
    layer.trainable=False
for layer in base_model.layers[-50:]:
    layer.trainable=True
# compile with current lr
model.compile(optimizer=optimizers.Adam(learning_rate=current_lr), loss=loss_fn, metrics=['accuracy'])

train_sac_phase("finetune", PHASE2_EPOCHS, allow_random_warmup=False)

# -------------------- SWA checkpoint averaging (post-training) --------------------
def average_checkpoints(checkpoint_paths, out_path):
    if not checkpoint_paths:
        return None
    models = []
    # load full-model .keras files if present; otherwise attempt weights (skip mismatch)
    for p in checkpoint_paths:
        try:
            m = keras.models.load_model(p)
            models.append(m)
        except Exception:
            try:
                # try weights only: create clone and set weights if possible
                tmp = keras.models.clone_model(model)
                tmp.build((None,)+IMG_SIZE+(3,))
                tmp.load_weights(p)
                models.append(tmp)
            except Exception:
                print("Warning: couldn't load checkpoint for SWA:", p)
    if not models:
        return None
    ws = [m.get_weights() for m in models]
    avg = []
    for layer_idx in range(len(ws[0])):
        layer_vals = [w[layer_idx] for w in ws]
        avg_layer = np.mean(layer_vals, axis=0)
        avg.append(avg_layer)
    m0 = keras.models.clone_model(models[0])
    m0.build((None,)+IMG_SIZE+(3,))
    m0.set_weights(avg)
    m0.save(out_path)
    return out_path

# average last 5 ckpts
ckpts = sorted(glob.glob(os.path.join(CKPT_DIR, "*.keras")), key=os.path.getmtime, reverse=True)[:5]
if ckpts:
    swa_path = os.path.join(RUN_DIR, "swa_avg.keras")
    avg_p = average_checkpoints(ckpts, swa_path)
    if avg_p:
        print("SWA averaged model saved to:", avg_p)

# -------------------- Final evaluation --------------------
print("Final evaluation on test set...")

if len(test_gen) == 0:
    print("No test images found in test generator. Skipping evaluation.")
    test_acc = 0.0
    y_true = np.array([])
    y_pred = np.array([])
    mis_rows = []
    mis_csv = ""
else:
    # choose model for final eval: swa if exists, else best, else final
    eval_model_path = None
    if 'swa_path' in locals() and swa_path and os.path.exists(swa_path):
        eval_model_path = swa_path
    elif os.path.exists(BEST_MODEL_PATH):
        eval_model_path = BEST_MODEL_PATH
    else:
        eval_model_path = None

    if eval_model_path:
        print("Loading evaluation model:", eval_model_path)
        eval_model = keras.models.load_model(eval_model_path)
    else:
        eval_model = model

    # optional Test Time Augmentation (TTA)
    def tta_predict_model(m, gen, tta_iters=5):
        # predicts over the generator's filepaths applying random flips/brightness; simple implementation
        preds = []
        for i in range(len(gen)):
            x_batch, y_batch = next(gen)
            batch_preds = np.zeros((x_batch.shape[0], num_classes), dtype=np.float32)
            for t in range(tta_iters):
                xt = x_batch.copy()
                # apply small random transforms using numpy (already preprocessed)
                if np.random.rand() < 0.5:
                    xt = xt[:, :, ::-1, :]  # horizontal flip (on preprocessed array it's safe)
                p = m.predict(xt, verbose=0)
                batch_preds += p
            batch_preds /= float(tta_iters)
            preds.append(batch_preds)
        preds = np.vstack(preds)
        return preds

    # reset generator
    test_gen.reset()
    # use TTA=3 to not take too long
    y_prob = tta_predict_model(eval_model, test_gen, tta_iters=3)
    y_pred = np.argmax(y_prob, axis=1)
    y_true = np.array(test_gen.classes)
    test_acc = np.mean(y_pred == y_true)
    print("Test accuracy (TTA):", test_acc)

    # classification report
    try:
        report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
        report_df = pd.DataFrame(report).transpose()
        report_df.to_csv(PER_CLASS_CSV)
        print("Saved per-class report:", PER_CLASS_CSV)
    except Exception as e:
        print("Warning: failed to save classification report:", e)

    # misclassified
    mis_rows = []
    filenames = test_gen.filepaths
    for i, fp in enumerate(filenames):
        if i >= len(y_pred): break
        t = int(y_true[i]); p = int(y_pred[i])
        if p != t:
            mis_rows.append({'filepath': fp, 'true': class_names[t], 'pred': class_names[p]})
            dst = os.path.join(MISCLASSIFIED_DIR, f"true_{class_names[t]}_pred_{class_names[p]}_{os.path.basename(fp)}")
            try:
                shutil.copy2(fp, dst)
            except Exception:
                pass
    mis_csv = os.path.join(RUN_DIR, "misclassified_list.csv")
    try:
        pd.DataFrame(mis_rows).to_csv(mis_csv, index=False)
        print("Saved misclassified examples to:", MISCLASSIFIED_DIR)
        print("Misclassified list saved to:", mis_csv)
    except Exception:
        mis_csv = ""

    # confusion matrix
    try:
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(10,8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
        plt.title("Confusion Matrix â€” MobileNetV2 + SAC (All Features)")
        plt.xlabel("Predicted"); plt.ylabel("Actual")
        plt.show()
    except Exception as e:
        print("Warning: couldn't plot confusion matrix:", e)

# -------------------- append summary and save final model --------------------
def save_final_and_append_results(test_acc, mis_csv):
    try:
        model.save(FINAL_MODEL_PATH)
    except Exception as e:
        print("Warning: failed to save final model:", e)
    final_row = {
        'timestamp': datetime.datetime.now().isoformat(),
        'method': 'MobileNetV2+SAC_allfeatures',
        'test_accuracy': float(test_acc),
        'num_classes': int(num_classes),
        'best_val_accuracy': float(best_val_acc),
        'final_model_path': FINAL_MODEL_PATH,
        'best_model_path': BEST_MODEL_PATH,
        'log_csv': LOG_CSV,
        'per_class_report': PER_CLASS_CSV,
        'misclassified_dir': MISCLASSIFIED_DIR,
        'misclassified_list_csv': mis_csv,
        'run_dir': RUN_DIR
    }
    try:
        if not os.path.exists(RESULTS_MASTER):
            pd.DataFrame([final_row]).to_csv(RESULTS_MASTER, index=False)
        else:
            dfm = pd.read_csv(RESULTS_MASTER)
            dfm = pd.concat([dfm, pd.DataFrame([final_row])], ignore_index=True)
            dfm.to_csv(RESULTS_MASTER, index=False)
        print("Appended run summary to:", RESULTS_MASTER)
    except Exception as e:
        print("Warning: couldn't append to results_master:", e)

save_final_and_append_results(test_acc, mis_csv if len(mis_rows)>0 else "")
try:
    model.save(BEST_MODEL_PATH)
except Exception as e:
    print("Warning: couldn't save BEST_MODEL_PATH:", e)
print("Saved final model to:", FINAL_MODEL_PATH)
print("Saved best model to:", BEST_MODEL_PATH)
print("Saved log to:", LOG_CSV)
print("Run directory:", RUN_DIR)
```
