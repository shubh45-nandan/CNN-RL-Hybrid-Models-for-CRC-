```python


# -------------------- Colab Drive mount (robust) --------------------
MOUNT_POINT = '/content/drive'
try:
    from google.colab import drive
    try:
        drive.mount(MOUNT_POINT, force_remount=True)
    except Exception as e:
        import os, shutil
        try:
            if os.path.exists(MOUNT_POINT):
                try:
                    shutil.rmtree(MOUNT_POINT)
                except Exception:
                    pass
            os.makedirs(MOUNT_POINT, exist_ok=True)
            drive.mount(MOUNT_POINT, force_remount=True)
        except Exception as e2:
            raise RuntimeError(f"Drive mount failed: {e} / {e2}")
except Exception as e:
    print("Warning: google.colab.drive not available. Are you running this in Colab?", e)

# -------------------- imports --------------------
import os, glob, shutil, random, zipfile, math, json, datetime, itertools, sys, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, optimizers, regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

print("✅ Libraries ready, TF version:", tf.__version__)

# -------------------- USER EDIT --------------------
ZIP_IN_DRIVE = "/content/drive/MyDrive/archive.zip"   # <<-- edit to your zip
PROJECT_DIR  = "/content/drive/MyDrive/RL_CNN_Research/mobilenetv2_ppo_full"  # top-level project folder
os.makedirs(PROJECT_DIR, exist_ok=True)
RESULTS_MASTER = os.path.join(PROJECT_DIR, "results_master.csv")
# ---------------------------------------------------

# -------------------- per-run folder --------------------
run_ts = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
rand_suf = f"{np.random.randint(1000,9999)}"
RUN_ID = f"run_{run_ts}_{rand_suf}"
RUN_DIR = os.path.join(PROJECT_DIR, RUN_ID)
os.makedirs(RUN_DIR, exist_ok=True)

# redirect commonly used paths into RUN_DIR
CKPT_DIR = os.path.join(RUN_DIR, "checkpoints"); os.makedirs(CKPT_DIR, exist_ok=True)
LOG_CSV = os.path.join(RUN_DIR, "ppo_onpolicy_log.csv")
BEST_MODEL_PATH = os.path.join(RUN_DIR, "mobilenetv2_ppo_onpolicy_best.keras")
FINAL_MODEL_PATH = os.path.join(RUN_DIR, "mobilenetv2_ppo_onpolicy_final.keras")
PER_CLASS_CSV = os.path.join(RUN_DIR, "per_class_report.csv")
MISCLASSIFIED_DIR = os.path.join(RUN_DIR, "misclassified"); os.makedirs(MISCLASSIFIED_DIR, exist_ok=True)

# record run info
with open(os.path.join(RUN_DIR, "run_info.txt"), "w") as f:
    f.write(f"RUN_ID: {RUN_ID}\nPROJECT_DIR: {PROJECT_DIR}\nZIP_IN_DRIVE: {ZIP_IN_DRIVE}\nTIMESTAMP: {run_ts}\n")

# -------------------- config & seeds --------------------
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)
VAL_RATIO = 0.20; TEST_RATIO = 0.15
IMG_SIZE = (224,224); BATCH_SIZE = 32
EXTS = (".jpg",".jpeg",".png",".bmp",".tif",".tiff")

# -------------------- unzip dataset --------------------
EXTRACT_DIR = "/content/Extracted_Mobile_PPO_Full"
if os.path.exists(EXTRACT_DIR): shutil.rmtree(EXTRACT_DIR)
os.makedirs(EXTRACT_DIR, exist_ok=True)

if not os.path.exists(ZIP_IN_DRIVE):
    raise FileNotFoundError(f"ZIP not found at {ZIP_IN_DRIVE}. Did you set ZIP_IN_DRIVE correctly and mount Drive?")
print("Unzipping... (this may take a while for big zips)")
with zipfile.ZipFile(ZIP_IN_DRIVE, 'r') as zf:
    zf.extractall(EXTRACT_DIR)
print("Unzipped.")

def list_images_recursive(folder):
    out=[]
    for r,_,files in os.walk(folder):
        for f in files:
            if os.path.splitext(f)[1].lower() in EXTS:
                out.append(os.path.join(r,f))
    return sorted(out)

def find_dataset_root(base):
    candidates=[]
    for root, dirs, _ in os.walk(base):
        if not dirs: continue
        score=0
        for d in dirs:
            dpath=os.path.join(root,d)
            for f in os.listdir(dpath):
                if os.path.splitext(f)[1].lower() in EXTS:
                    score+=1; break
        if score>0: candidates.append((root,score))
    if candidates:
        candidates.sort(key=lambda x:x[1], reverse=True); return candidates[0][0]
    return base

DATA_ROOT = find_dataset_root(EXTRACT_DIR)
print("DATA_ROOT:", DATA_ROOT)

# -------------------- map classes & split --------------------
class_to_files={}
for entry in sorted(os.listdir(DATA_ROOT)):
    p=os.path.join(DATA_ROOT, entry)
    if os.path.isdir(p):
        imgs=list_images_recursive(p)
        if imgs: class_to_files[entry]=imgs
if not class_to_files:
    # fallback scan deeper
    for r,dirs,files in os.walk(DATA_ROOT):
        for d in dirs:
            p=os.path.join(r,d)
            imgs=list_images_recursive(p)
            if imgs: class_to_files[d]=imgs

available_classes = sorted(class_to_files.keys())
print("Classes and counts:", [(c,len(class_to_files[c])) for c in available_classes])

SPLIT_ROOT = "/content/mobilenet_ppo_full_split"
if os.path.exists(SPLIT_ROOT): shutil.rmtree(SPLIT_ROOT)
TRAIN_DIR = os.path.join(SPLIT_ROOT,"train"); VAL_DIR = os.path.join(SPLIT_ROOT,"val"); TEST_DIR = os.path.join(SPLIT_ROOT,"test")
os.makedirs(TRAIN_DIR, exist_ok=True); os.makedirs(VAL_DIR, exist_ok=True); os.makedirs(TEST_DIR, exist_ok=True)

def allocate_counts(n,val_ratio=VAL_RATIO,test_ratio=TEST_RATIO):
    if n<=0: return (0,0,0)
    if n==1: return (1,0,0)
    if n==2: return (1,0,1)
    test_n=max(1,int(round(n*test_ratio)))
    val_n=max(1,int(round(n*val_ratio)))
    if test_n+val_n>=n:
        # ensure at least one train
        if n>=3:
            val_n=1; test_n=1
        else:
            # fallback safe split
            val_n=0; test_n=1
    train_n=n-val_n-test_n; return (train_n,val_n,test_n)

rng=np.random.default_rng(SEED)
tot={"train":0,"val":0,"test":0}
for cls, files in class_to_files.items():
    files = sorted(files); n=len(files); idx=np.arange(n); rng.shuffle(idx)
    tr_n, v_n, te_n = allocate_counts(n)
    tr_idx=set(idx[:tr_n]); v_idx=set(idx[tr_n:tr_n+v_n]); te_idx=set(idx[tr_n+v_n:tr_n+v_n+te_n])
    os.makedirs(os.path.join(TRAIN_DIR,cls), exist_ok=True)
    os.makedirs(os.path.join(VAL_DIR,cls), exist_ok=True)
    os.makedirs(os.path.join(TEST_DIR,cls), exist_ok=True)
    for i in idx:
        src=files[i]; basename=os.path.basename(src)
        if i in tr_idx: dst=os.path.join(TRAIN_DIR,cls,basename)
        elif i in v_idx: dst=os.path.join(VAL_DIR,cls,basename)
        else: dst=os.path.join(TEST_DIR,cls,basename)
        try:
            shutil.copy2(src,dst)
        except Exception:
            # if copy fails, still continue
            pass
    tot['train']+=len(tr_idx); tot['val']+=len(v_idx); tot['test']+=len(te_idx)
print("Split done:", tot)

# -------------------- datagens --------------------
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input,
                                   rotation_range=20, width_shift_range=0.1, height_shift_range=0.1,
                                   zoom_range=0.15, horizontal_flip=True)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_directory(TRAIN_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=True)
val_gen   = val_datagen.flow_from_directory(VAL_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False)
test_gen  = val_datagen.flow_from_directory(TEST_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False)
class_names = list(train_gen.class_indices.keys()); num_classes=len(class_names)
print("Detected classes:", class_names)

# save class_indices mapping for reproducibility
try:
    json.dump(train_gen.class_indices, open(os.path.join(RUN_DIR,'class_indices.json'),'w'))
except Exception:
    pass

# -------------------- build MobileNetV2 classifier --------------------
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=IMG_SIZE+(3,))
base_model.trainable=False
inputs = keras.Input(shape=IMG_SIZE+(3,))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation='softmax')(x)
model = keras.Model(inputs, outputs)
INITIAL_LR = 1e-4
model.compile(optimizer=optimizers.Adam(learning_rate=INITIAL_LR), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# -------------------- Full PPO (on-policy) hyperparams & options --------------------
ACTIONS = [0.5, 0.75, 1.0, 1.25, 1.5]
N_ACTIONS = len(ACTIONS)
STATE_DIM = 3
ROLLOUT_STEPS = 6        # epochs per rollout
GAE_LAMBDA = 0.95
GAMMA = 0.99
ENTROPY_COEF = 0.01
ACTOR_LR = 3e-4
CRITIC_LR = 1e-3
PPO_EPOCHS = 4
MINI_BATCH = 32
GRAD_CLIP_NORM = 0.5
WARMUP_ROLLOUTS = 2
PHASE1_EPOCHS = 15
PHASE2_EPOCHS = 25
CLIP_EPS = 0.2

# Optional/advanced features toggles and hyperparams
USE_VALUE_CLIP = True            # clip value function updates
USE_KL_PENALTY = True            # use KL penalty variant (alternatively to clipped objective)
KL_TARGET = 0.01                 # target KL for adaptive penalty or early stopping
KL_BETA = 1.0                    # initial KL penalty coefficient
ADAPT_KL = True                  # adapt KL_BETA to match KL_TARGET
USE_KL_EARLY_STOP = True         # stop epoch updates if KL > 1.5 * KL_TARGET
ENTROPY_ANNEAL = True            # linearly decay entropy coef over time
ENTROPY_START = 0.02
ENTROPY_END = 0.0
CLIP_ANNEAL = True               # anneal clip eps over training
CLIP_START = 0.2
CLIP_END = 0.05
REWARD_NORM = True               # normalize per-step rewards with running mean/std
OBS_NORM = True                  # normalize states (running mean/std)
VALUE_COEF = 0.5
MAX_GRAD_NORM = 0.5

# -------------------- actor & critic networks --------------------
def build_actor():
    inp = keras.Input(shape=(STATE_DIM,))
    x = layers.Dense(128, activation='relu')(inp)
    x = layers.Dense(128, activation='relu')(x)
    logits = layers.Dense(N_ACTIONS)(x)
    return keras.Model(inp, logits)

def build_critic():
    inp = keras.Input(shape=(STATE_DIM,))
    x = layers.Dense(128, activation='relu')(inp)
    x = layers.Dense(128, activation='relu')(x)
    v = layers.Dense(1)(x)
    return keras.Model(inp, v)

actor = build_actor()
critic = build_critic()
actor_opt = optimizers.Adam(learning_rate=ACTOR_LR)
critic_opt = optimizers.Adam(learning_rate=CRITIC_LR)

# -------------------- running mean/std for observation & reward normalization --------------------
class RunningMeanStd:
    def __init__(self, shape=(), eps=1e-4):
        self.mean = np.zeros(shape, 'float64')
        self.var = np.ones(shape, 'float64')
        self.count = eps

    def update(self, x):
        x = np.array(x, dtype=np.float64)
        # defensive handling: allow scalars, 1D or 2D arrays
        if x.ndim == 0:
            x = x.reshape(1)
        batch_mean = np.mean(x, axis=0)
        batch_var = np.var(x, axis=0)
        batch_count = x.shape[0]
        self._update_from_moments(batch_mean, batch_var, batch_count)

    def _update_from_moments(self, batch_mean, batch_var, batch_count):
        delta = batch_mean - self.mean
        tot_count = self.count + batch_count
        new_mean = self.mean + delta * batch_count / tot_count
        m_a = self.var * (self.count)
        m_b = batch_var * (batch_count)
        M2 = m_a + m_b + np.square(delta) * (self.count * batch_count / tot_count)
        new_var = M2 / (tot_count)
        self.mean = new_mean
        self.var = new_var
        self.count = tot_count

    def normalize(self, x):
        x = np.array(x, dtype=np.float64)
        return (x - self.mean) / (np.sqrt(self.var) + 1e-8)

obs_rms = RunningMeanStd(shape=(STATE_DIM,)) if OBS_NORM else None
rew_rms = RunningMeanStd(shape=()) if REWARD_NORM else None

# -------------------- helpers & checkpoint saver --------------------
def set_optimizer_lr(optim, lr):
    try:
        # prefer assign if available (tf.Variable)
        optim.learning_rate.assign(lr)
    except Exception:
        try:
            tf.keras.backend.set_value(optim.learning_rate, lr)
        except Exception:
            # last-resort: set attribute (may not persist)
            try: optim.learning_rate = lr
            except Exception: pass

def state_from(val_acc, prev_val_acc_local, lr):
    delta = float(val_acc - prev_val_acc_local)
    lmin, lmax = math.log10(1e-6), math.log10(1e-3)
    lrn = (math.log10(max(lr,1e-12)) - lmin) / (lmax - lmin + 1e-8)
    return np.array([float(val_acc), float(delta), float(lrn)], dtype=np.float32)

# Corrected weights checkpoint saver (filename must end with .weights.h5)
def save_checkpoint_weights(model_obj, phase, update_idx, keep_last=3):
    fname = os.path.join(CKPT_DIR, f"{phase}_update{update_idx}.weights.h5")
    try:
        model_obj.save_weights(fname)
    except Exception as e:
        print("Warning: failed to save weights:", e)
    pattern = os.path.join(CKPT_DIR, f"{phase}_update*.weights.h5")
    all_ck = sorted([p for p in glob.glob(pattern)], key=os.path.getmtime, reverse=True)
    for old in all_ck[keep_last:]:
        try: os.remove(old)
        except: pass
    return fname

# -------------------- PPO utilities --------------------
def sample_action_and_logprob(state_np):
    # state_np shape: (1,STATE_DIM) or (STATE_DIM,)
    st = tf.convert_to_tensor(state_np, dtype=tf.float32)
    if len(st.shape) == 1:
        st = tf.expand_dims(st, 0)
    logits = actor(st)
    logp = tf.nn.log_softmax(logits)
    probs = tf.nn.softmax(logits)
    idx = int(tf.squeeze(tf.random.categorical(logits, 1)).numpy())
    logp_chosen = float(logp[0, idx].numpy())
    return int(idx), float(logp_chosen)

def compute_gae(rewards, values, last_value, dones, gamma=GAMMA, lam=GAE_LAMBDA):
    T = len(rewards)
    adv = np.zeros(T, dtype=np.float32)
    last_gae = 0.0
    for t in reversed(range(T)):
        nonterminal = 1.0 - dones[t]
        next_val = last_value if t==T-1 else values[t+1]
        delta = rewards[t] + gamma * next_val * nonterminal - values[t]
        last_gae = delta + gamma * lam * nonterminal * last_gae
        adv[t] = last_gae
    returns = adv + values
    return adv, returns

@tf.function
def ppo_update_step(states, actions, old_logps, advantages, returns, old_values, clip_eps, entropy_coef, kl_beta, use_value_clip):
    with tf.GradientTape() as tape_a, tf.GradientTape() as tape_v:
        logits = actor(states)
        logp_all = tf.nn.log_softmax(logits)
        actions_onehot = tf.one_hot(actions, N_ACTIONS, dtype=tf.float32)
        new_logp = tf.reduce_sum(logp_all * actions_onehot, axis=1, keepdims=True)

        ratio = tf.exp(new_logp - old_logps)
        adv = advantages
        surr1 = ratio * adv
        surr2 = tf.clip_by_value(ratio, 1.0 - clip_eps, 1.0 + clip_eps) * adv
        policy_loss_clip = -tf.reduce_mean(tf.minimum(surr1, surr2))

        approx_kl = tf.reduce_mean(old_logps - new_logp)

        entropy = -tf.reduce_mean(tf.reduce_sum(tf.exp(logp_all) * logp_all, axis=1))

        total_policy_loss = policy_loss_clip - entropy_coef * entropy

        v_pred = critic(states)
        value_loss = tf.reduce_mean(tf.square(returns - v_pred))
        if use_value_clip:
            v_pred_old = old_values
            v_pred_clipped = v_pred_old + tf.clip_by_value(v_pred - v_pred_old, -clip_eps, clip_eps)
            value_loss_unclipped = tf.square(returns - v_pred)
            value_loss_clipped = tf.square(returns - v_pred_clipped)
            value_loss = 0.5 * tf.reduce_mean(tf.maximum(value_loss_unclipped, value_loss_clipped))

    grads_a = tape_a.gradient(total_policy_loss, actor.trainable_variables)
    grads_v = tape_v.gradient(value_loss, critic.trainable_variables)
    grads_a, _ = tf.clip_by_global_norm(grads_a, MAX_GRAD_NORM)
    grads_v, _ = tf.clip_by_global_norm(grads_v, MAX_GRAD_NORM)
    actor_opt.apply_gradients(zip(grads_a, actor.trainable_variables))
    critic_opt.apply_gradients(zip(grads_v, critic.trainable_variables))

    return policy_loss_clip, value_loss, entropy, approx_kl

# -------------------- rollout collection & updates --------------------
def collect_rollout(use_random_policy=False):
    global current_lr, prev_val_acc
    states=[]; actions=[]; logps=[]; rewards=[]; values=[]; dones=[]
    # keep a local copy of prev_val_acc for computing states during this rollout
    prev_val_local = prev_val_acc
    last_val_acc = prev_val_acc

    # protective steps per epoch
    train_steps = max(1, len(train_gen))
    val_steps = max(1, len(val_gen))

    for step in range(ROLLOUT_STEPS):
        # train for one epoch (intentionally repeated per rollout step)
        h = model.fit(train_gen, validation_data=val_gen, epochs=1, verbose=1,
                      steps_per_epoch=train_steps, validation_steps=val_steps)
        val_acc = float(h.history.get('val_accuracy', [0.0])[-1])
        val_loss = float(h.history.get('val_loss', [0.0])[-1])

        # state is computed from current val_acc and previous val_acc (prev_val_local)
        s = state_from(val_acc, prev_val_local, current_lr)
        # update running obs stats BEFORE policy sample (so actor sees normalized states)
        if obs_rms is not None:
            try:
                obs_rms.update(s.reshape(1,-1))
            except Exception:
                obs_rms.update(s)
            s = obs_rms.normalize(s)
        s_tf = s.reshape(1,-1).astype(np.float32)

        if use_random_policy:
            action_idx = np.random.randint(0, N_ACTIONS)
            logp_chosen = math.log(1.0 / N_ACTIONS)
        else:
            action_idx, logp_chosen = sample_action_and_logprob(s_tf)

        action_multiplier = ACTIONS[action_idx]
        new_lr = float(np.clip(current_lr * action_multiplier, 1e-6, 1e-3))
        set_optimizer_lr(model.optimizer, new_lr)
        current_lr = new_lr

        # reward semantics: absolute val_acc or delta depending on preference
        reward = float(val_acc)
        # if you'd rather use improvement as reward, uncomment below:
        # reward = float(val_acc - prev_val_local)

        if rew_rms is not None:
            try:
                rew_rms.update(np.array([reward]))
                reward = float(rew_rms.normalize(np.array([reward]))[0])
            except Exception:
                pass

        done = 0.0
        v = float(critic(s_tf).numpy()[0,0])

        states.append(s); actions.append(action_idx); logps.append(logp_chosen)
        rewards.append(reward); values.append(v); dones.append(done)

        # set prev_val_local for next step (we want delta next step to be val_acc - this val_acc)
        prev_val_local = val_acc
        last_val_acc = val_acc

    # compute last value using the most recent val_acc and prev_val_local
    last_state = state_from(last_val_acc, prev_val_local, current_lr).reshape(1,-1).astype(np.float32)
    if obs_rms is not None:
        last_state = obs_rms.normalize(last_state)
    last_value = float(critic(last_state).numpy()[0,0])

    # update global prev_val_acc for outer loop tracking
    prev_val_acc = last_val_acc

    return {
        'states': np.array(states, dtype=np.float32),
        'actions': np.array(actions, dtype=np.int32),
        'logps': np.array(logps, dtype=np.float32).reshape(-1,1),
        'rewards': np.array(rewards, dtype=np.float32),
        'values': np.array(values, dtype=np.float32),
        'dones': np.array(dones, dtype=np.float32),
        'last_value': last_value,
        'mean_val_acc': float(np.mean([float(x) for x in rewards])) if len(rewards)>0 else 0.0,
        'chosen_actions': [ACTIONS[a] for a in actions]
    }

def update_from_rollout(rollout, clip_eps_current, entropy_coef_current, kl_beta_current):
    states = rollout['states']; actions = rollout['actions']; rewards = rollout['rewards']
    values = rollout['values']; dones = rollout['dones']; last_value = rollout['last_value']

    if len(rewards)==0:
        return 0.0, 0.0, 0.0, 0.0, kl_beta_current

    advantages, returns = compute_gae(list(rewards), list(values), last_value, list(dones), GAMMA, GAE_LAMBDA)
    # normalize advantages
    adv_mean = advantages.mean(); adv_std = advantages.std() + 1e-8
    advantages = ((advantages - adv_mean) / adv_std).reshape(-1,1).astype(np.float32)
    returns = returns.reshape(-1,1).astype(np.float32)

    old_logps = rollout['logps'].astype(np.float32)
    old_values = values.reshape(-1,1).astype(np.float32)

    N = len(states); idxs = np.arange(N)
    p_loss = v_loss = entropy = 0.0
    mean_kl = 0.0
    num_updates = 0
    for ep in range(PPO_EPOCHS):
        np.random.shuffle(idxs)
        for start in range(0, N, MINI_BATCH):
            mb_idx = idxs[start:start+MINI_BATCH]
            if len(mb_idx)==0: continue
            mb_states = tf.convert_to_tensor(states[mb_idx], dtype=tf.float32)
            mb_actions = tf.convert_to_tensor(actions[mb_idx], dtype=tf.int32)
            mb_old_logps = tf.convert_to_tensor(old_logps[mb_idx], dtype=tf.float32)
            mb_adv = tf.convert_to_tensor(advantages[mb_idx], dtype=tf.float32)
            mb_ret = tf.convert_to_tensor(returns[mb_idx], dtype=tf.float32)
            mb_old_vals = tf.convert_to_tensor(old_values[mb_idx], dtype=tf.float32)

            pl, vl, ent, approx_kl = ppo_update_step(mb_states, mb_actions, mb_old_logps, mb_adv, mb_ret, mb_old_vals, clip_eps_current, entropy_coef_current, kl_beta_current, USE_VALUE_CLIP)
            p_loss += float(pl.numpy()); v_loss += float(vl.numpy()); entropy += float(ent.numpy()); mean_kl += float(approx_kl.numpy())
            num_updates += 1

        # optional KL early stop (use mean KL across minibatches so far)
        if USE_KL_EARLY_STOP and num_updates>0:
            mean_kl_ep = mean_kl / num_updates
            if mean_kl_ep > 1.5 * KL_TARGET:
                print(f"Early stopping PPO epochs due to large KL: {mean_kl_ep:.6f}")
                break

    if num_updates > 0:
        p_loss /= num_updates; v_loss /= num_updates; entropy /= num_updates; mean_kl /= num_updates

    # adapt KL beta if requested
    if ADAPT_KL and USE_KL_PENALTY:
        if mean_kl > KL_TARGET * 1.5:
            kl_beta_current *= 1.5
        elif mean_kl < KL_TARGET / 1.5:
            kl_beta_current *= 0.67
    return p_loss, v_loss, entropy, mean_kl, kl_beta_current

# -------------------- training loop using rollouts --------------------
# prepare log CSV
if not os.path.exists(LOG_CSV):
    pd.DataFrame(columns=['timestamp','phase','update_idx','mean_val_acc','lr','chosen_actions','avg_reward','policy_loss','value_loss','entropy','kl','clip_eps','entropy_coef','kl_beta']).to_csv(LOG_CSV,index=False)

best_val_acc = -1.0
current_lr = INITIAL_LR
prev_val_acc = 0.0
clip_eps_current = CLIP_START if CLIP_ANNEAL else CLIP_EPS
entropy_coef_current = ENTROPY_START if ENTROPY_ANNEAL else ENTROPY_COEF
kl_beta_current = KL_BETA

# save helper
def save_final_and_append_results(test_acc, mis_csv):
    # final save and append summary to results_master
    try:
        model.save(FINAL_MODEL_PATH)
    except Exception as e:
        print("Warning: failed to save final model:", e)
    final_row = {
        'timestamp': datetime.datetime.now().isoformat(),
        'method': 'MobileNetV2+PPO_onpolicy_full',
        'test_accuracy': float(test_acc),
        'num_classes': int(num_classes),
        'best_val_accuracy': float(best_val_acc),
        'final_model_path': FINAL_MODEL_PATH,
        'best_model_path': BEST_MODEL_PATH,
        'log_csv': LOG_CSV,
        'per_class_report': PER_CLASS_CSV,
        'misclassified_dir': MISCLASSIFIED_DIR,
        'misclassified_list_csv': mis_csv,
        'run_dir': RUN_DIR
    }
    try:
        if not os.path.exists(RESULTS_MASTER):
            pd.DataFrame([final_row]).to_csv(RESULTS_MASTER, index=False)
        else:
            dfm = pd.read_csv(RESULTS_MASTER)
            dfm = pd.concat([dfm, pd.DataFrame([final_row])], ignore_index=True)
            dfm.to_csv(RESULTS_MASTER, index=False)
        print("Appended run summary to:", RESULTS_MASTER)
    except Exception as e:
        print("Warning: couldn't append to results_master:", e)

# main training phase wrapped in try/except to save last state on crash

def train_phase_with_ppo(phase_name, total_epochs, allow_random_warmup=True):
    global prev_val_acc, best_val_acc, current_lr, clip_eps_current, entropy_coef_current, kl_beta_current
    epochs_done = 0; update_idx = 0
    try:
        while epochs_done < total_epochs:
            use_random = allow_random_warmup and (update_idx < WARMUP_ROLLOUTS)
            print(f"\n=== {phase_name} rollout {update_idx+1} | random_rollout={use_random} ===")
            rollout = collect_rollout(use_random_policy=use_random)
            epochs_done += len(rollout['rewards'])

            # anneal clip & entropy
            if CLIP_ANNEAL:
                frac = min(1.0, float(epochs_done) / float(total_epochs))
                clip_eps_current = CLIP_START + frac * (CLIP_END - CLIP_START)
            if ENTROPY_ANNEAL:
                frac = min(1.0, float(epochs_done) / float(total_epochs))
                entropy_coef_current = ENTROPY_START + frac * (ENTROPY_END - ENTROPY_START)

            p_loss, v_loss, ent, mean_kl, kl_beta_current = update_from_rollout(rollout, clip_eps_current, entropy_coef_current, kl_beta_current)

            # if using KL penalty, add it to policy loss log
            if USE_KL_PENALTY:
                p_loss_logged = p_loss + kl_beta_current * mean_kl
            else:
                p_loss_logged = p_loss

            # save intermediate weights
            save_checkpoint_weights(model, phase_name, update_idx, keep_last=3)

            mean_val = rollout['mean_val_acc']
            if mean_val > best_val_acc:
                best_val_acc = mean_val
                try:
                    model.save(BEST_MODEL_PATH)
                    print(f"Saved new best model (mean_val_acc={mean_val:.4f})")
                except Exception as e:
                    print("Warning: failed to save best model:", e)

            # log
            row = {
                'timestamp': str(datetime.datetime.now()),
                'phase': phase_name,
                'update_idx': update_idx,
                'mean_val_acc': float(mean_val),
                'lr': float(current_lr),
                'chosen_actions': json.dumps(rollout['chosen_actions']),
                'avg_reward': float(np.mean(rollout['rewards'])) if len(rollout['rewards'])>0 else 0.0,
                'policy_loss': float(p_loss_logged),
                'value_loss': float(v_loss),
                'entropy': float(ent),
                'kl': float(mean_kl),
                'clip_eps': float(clip_eps_current),
                'entropy_coef': float(entropy_coef_current),
                'kl_beta': float(kl_beta_current)
            }
            df = pd.read_csv(LOG_CSV)
            df = pd.concat([df, pd.DataFrame([row])], ignore_index=True)
            df.to_csv(LOG_CSV, index=False)

            print(f"[{phase_name}] rollout {update_idx} | mean_val_acc={mean_val:.4f} | lr={current_lr:.6f} | actions={rollout['chosen_actions']} | kl={mean_kl:.6f} | clip={clip_eps_current:.4f} | ent_coef={entropy_coef_current:.4f}")
            update_idx += 1
    except Exception as e:
        print(f"Exception during {phase_name} training: {e}")
        # try to save model/weights so progress is not lost
        try:
            model.save(os.path.join(RUN_DIR, f"{phase_name}_interrupted.keras"))
        except Exception:
            try:
                save_checkpoint_weights(model, f"{phase_name}_interrupted", update_idx, keep_last=3)
            except Exception:
                pass
        raise

# -------------------- Run phases --------------------
print("=== Phase 1: frozen training (PPO controls LR) ===")
train_phase_with_ppo("frozen", PHASE1_EPOCHS, allow_random_warmup=True)

print("=== Phase 2: fine-tune (unfreeze top layers) ===")
base_model.trainable=True
for layer in base_model.layers[:-50]:
    layer.trainable=False
for layer in base_model.layers[-50:]:
    layer.trainable=True
model.compile(optimizer=optimizers.Adam(learning_rate=current_lr), loss='categorical_crossentropy', metrics=['accuracy'])

train_phase_with_ppo("finetune", PHASE2_EPOCHS, allow_random_warmup=False)

# -------------------- Final evaluation, per-class logging, append to master --------------------
print("Final evaluation on test set...")

if len(test_gen) == 0:
    print("No test images found in test generator. Skipping evaluation.")
    test_acc = 0.0
    y_true = np.array([])
    y_pred = np.array([])
    mis_rows = []
    mis_csv = ""
else:
    test_res = model.evaluate(test_gen, verbose=1, steps=max(1, len(test_gen)))
    test_acc = float(test_res[1]) if len(test_res) > 1 else float(test_res[0])
    # predictions
    y_true = np.array(test_gen.classes)
    y_pred = np.argmax(model.predict(test_gen, verbose=0, steps=max(1, len(test_gen))), axis=1)

    # classification report + per-class csv
    try:
        report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
        report_df = pd.DataFrame(report).transpose()
        report_df.to_csv(PER_CLASS_CSV)
        print("Saved per-class report:", PER_CLASS_CSV)
    except Exception as e:
        print("Warning: failed to save classification report:", e)

    # Save per-class counts
    try:
        counts = pd.Series(y_true).value_counts().sort_index()
        counts.index = [class_names[i] for i in counts.index]
        counts.to_frame('count').to_csv(os.path.join(RUN_DIR, "per_class_counts.csv"))
    except Exception:
        pass

    # misclassification log (list of files mispredicted) + csv
    mis_rows = []
    filenames = test_gen.filepaths
    for i, fp in enumerate(filenames):
        if i >= len(y_pred): break
        t = int(y_true[i]); p = int(y_pred[i])
        if p != t:
            mis_rows.append({'filepath': fp, 'true': class_names[t], 'pred': class_names[p]})
            dst = os.path.join(MISCLASSIFIED_DIR, f"true_{class_names[t]}_pred_{class_names[p]}_{os.path.basename(fp)}")
            try:
                shutil.copy2(fp, dst)
            except Exception:
                pass
    mis_csv = os.path.join(RUN_DIR, "misclassified_list.csv")
    try:
        pd.DataFrame(mis_rows).to_csv(mis_csv, index=False)
        print("Saved misclassified examples to:", MISCLASSIFIED_DIR)
        print("Misclassified list saved to:", mis_csv)
    except Exception:
        mis_csv = ""

    # confusion matrix plot (guarded)
    try:
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(10,8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
        plt.title("Confusion Matrix — MobileNetV2 + PPO (Full Variant)")
        plt.xlabel("Predicted"); plt.ylabel("Actual")
        plt.show()
    except Exception as e:
        print("Warning: couldn't plot confusion matrix:", e)

# append to results_master and save final model
save_final_and_append_results(test_acc, mis_csv if len(mis_rows)>0 else "")
# save best again for convenience
try:
    model.save(BEST_MODEL_PATH)
except Exception as e:
    print("Warning: couldn't save BEST_MODEL_PATH:", e)
print("Saved final model to:", FINAL_MODEL_PATH)
print("Saved best model to:", BEST_MODEL_PATH)
print("Saved log to:", LOG_CSV)
print("Run directory:", RUN_DIR)  # debug this
```

```python
# === Test evaluation & summary (single-cell) ===
# Paste & run in the same Colab runtime after training finished.
# Adjust PROJECT_DIR or set RUN_DIR / TEST_DIR if needed.

import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
from IPython.display import Image, display
import seaborn as sns
import tensorflow as tf
from tensorflow import keras

# ---------- User-editable paths ----------
# Default points to the project folder used by the patched script
PROJECT_DIR = "/content/drive/MyDrive/RL_CNN_Research/mobilenetv2_ppo_full"
# Optionally override to a specific run folder:
RUN_DIR = None   # e.g. "/content/drive/MyDrive/.../run_2025..."
# Optionally specify test data folder if not in the usual split path:
TEST_DIR = None  # e.g. "/content/mobilenet_ppo_full_split/test"
# Optionally specify a model path (best or final .keras). If None, we auto-detect in RUN_DIR.
MODEL_PATH = None
# Output filenames (will be saved into RUN_DIR or PROJECT_DIR)
OUT_PROJECT_DIR = PROJECT_DIR
# -----------------------------------------

def find_latest_run(project_dir):
    if not os.path.exists(project_dir):
        return None
    runs = [p for p in glob.glob(os.path.join(project_dir, "run_*")) if os.path.isdir(p)]
    if not runs:
        return None
    runs.sort(key=os.path.getmtime, reverse=True)
    return runs[0]

# Resolve RUN_DIR
if RUN_DIR is None:
    RUN_DIR = find_latest_run(PROJECT_DIR)
    if RUN_DIR is None:
        raise FileNotFoundError(f"No run_* directory found under PROJECT_DIR={PROJECT_DIR}. Set RUN_DIR manually.")
print("Using RUN_DIR =", RUN_DIR)

# Resolve MODEL_PATH
if MODEL_PATH is None:
    # look for typical best/final model names in RUN_DIR
    cand = glob.glob(os.path.join(RUN_DIR, "*best*.keras")) + glob.glob(os.path.join(RUN_DIR, "*final*.keras"))
    cand = sorted(cand, key=os.path.getmtime, reverse=True)
    MODEL_PATH = cand[0] if cand else None

# Resolve TEST_DIR (if not provided, try common split path relative to RUN_DIR parent)
if TEST_DIR is None:
    parent = os.path.dirname(RUN_DIR)
    candidates = [
        os.path.join(parent, "mobilenet_ppo_full_split", "test"),
        os.path.join(parent, "mobilenet_ppo_full_split", "test"),
        os.path.join(parent, "mobilenet_ppo_full_split"),
        os.path.join(RUN_DIR, "test"),
        os.path.join(parent, "test"),
    ]
    # also search grandchildren for a folder named 'test'
    for p in glob.glob(os.path.join(parent, "**", "test"), recursive=True):
        if p not in candidates:
            candidates.append(p)
    TEST_DIR = next((p for p in candidates if p and os.path.exists(p)), None)

print("Auto-detected MODEL_PATH =", MODEL_PATH)
print("Auto-detected TEST_DIR  =", TEST_DIR)

# Choose output directory for summaries
OUT_DIR = RUN_DIR if RUN_DIR else OUT_PROJECT_DIR
os.makedirs(OUT_DIR, exist_ok=True)

# filenames
cm_path = os.path.join(OUT_DIR, "confusion_matrix.png")
per_class_csv = os.path.join(OUT_DIR, "per_class_report.csv")
summary_csv = os.path.join(OUT_DIR, "test_summary.csv")

# ------------- Load model -------------
if 'model' in globals() and isinstance(globals()['model'], keras.Model):
    print("Using in-memory model object.")
    model = globals()['model']
elif MODEL_PATH and os.path.exists(MODEL_PATH):
    print("Loading model from:", MODEL_PATH)
    model = keras.models.load_model(MODEL_PATH)
else:
    raise FileNotFoundError("No model found in memory and no MODEL_PATH available. Set MODEL_PATH or ensure model is defined.")

# ------------- Build test generator -------------
# If user provided a test_gen in the session, use it.
if 'test_gen' in globals():
    test_gen = globals()['test_gen']
    print("Using in-memory test_gen.")
elif TEST_DIR and os.path.exists(TEST_DIR):
    print("Creating test generator from TEST_DIR:", TEST_DIR)
    from tensorflow.keras.preprocessing.image import ImageDataGenerator
    from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
    IMG_SIZE = (224,224)
    BATCH_SIZE = 32
    val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
    test_gen = val_datagen.flow_from_directory(TEST_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
                                               class_mode='categorical', shuffle=False)
else:
    raise FileNotFoundError("No test generator found in memory and TEST_DIR not found. Set TEST_DIR or provide test_gen in the session.")

# ------------- Evaluate (model.evaluate) -------------
print("\nRunning model.evaluate on the test generator...")
# Guard steps to be at least 1
steps = max(1, len(test_gen))
eval_res = model.evaluate(test_gen, verbose=1, steps=steps)
if isinstance(eval_res, (list, tuple)) and len(eval_res) >= 2:
    test_loss = float(eval_res[0])
    test_acc = float(eval_res[1])
else:
    # If model returned single metric, treat it as accuracy if plausible
    try:
        test_acc = float(eval_res[0])
    except Exception:
        test_acc = float(np.nan)
    test_loss = float(np.nan)
print(f"Test loss: {test_loss}, Test accuracy: {test_acc}")

# ------------- Get y_true and y_pred -------------
# y_true
if hasattr(test_gen, "classes"):
    y_true = np.array(test_gen.classes)
else:
    # fallback: iterate once (consumes generator)
    y_true = []
    test_gen.reset()
    for _ in range(len(test_gen)):
        x_batch, y_batch = next(test_gen)
        if y_batch.ndim == 2:
            labels = np.argmax(y_batch, axis=1)
        else:
            labels = y_batch
        y_true.extend(labels.tolist())
    y_true = np.array(y_true)

# Predict (ensure generator is reset)
try:
    test_gen.reset()
except Exception:
    pass

y_prob = model.predict(test_gen, verbose=0, steps=max(1, len(test_gen)))
y_pred = np.argmax(y_prob, axis=1)

# class names
if 'class_names' in globals() and isinstance(globals()['class_names'], (list,tuple)):
    class_names = globals()['class_names']
else:
    # infer from generator
    class_names = list(test_gen.class_indices.keys())

# ------------- Per-class classification report & CSV -------------
print("\nClassification report (per-class):\n")
crep_text = classification_report(y_true, y_pred, target_names=class_names, zero_division=0)
print(crep_text)

report_dict = classification_report(y_true, y_pred, target_names=class_names, output_dict=True, zero_division=0)
report_df = pd.DataFrame(report_dict).transpose()
report_df.to_csv(per_class_csv)
print("Saved per-class report to:", per_class_csv)

# ------------- Precision/Recall/F1 arrays and summary CSV -------------
prec, rec, f1, sup = precision_recall_fscore_support(y_true, y_pred, labels=range(len(class_names)), zero_division=0)

summary_df = pd.DataFrame({
    "class": class_names,
    "precision": prec,
    "recall": rec,
    "f1": f1,
    "support": sup
})

# overall averages
micro_prec, micro_rec, micro_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='micro', zero_division=0)
macro_prec, macro_rec, macro_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='macro', zero_division=0)
weighted_prec, weighted_rec, weighted_f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=0)

summary_overall = pd.DataFrame([{
    "class": "MICRO_AVG", "precision": micro_prec, "recall": micro_rec, "f1": micro_f1, "support": len(y_true)
}, {
    "class": "MACRO_AVG", "precision": macro_prec, "recall": macro_rec, "f1": macro_f1, "support": len(y_true)
}, {
    "class": "WEIGHTED_AVG", "precision": weighted_prec, "recall": weighted_rec, "f1": weighted_f1, "support": len(y_true)
}])

summary_out = pd.concat([summary_df, summary_overall], ignore_index=True)
summary_out.to_csv(summary_csv, index=False)
print("Saved precision/recall/f1 summary to:", summary_csv)

# ------------- Confusion matrix plot & save -------------
cm = confusion_matrix(y_true, y_pred)
n = len(class_names)
fig_w = min(20, max(6, n * 0.5))
fig_h = min(20, max(4, n * 0.45))
fig, ax = plt.subplots(figsize=(fig_w, fig_h))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
            xticklabels=class_names, yticklabels=class_names, cbar=True)
ax.set_title("Confusion Matrix")
ax.set_xlabel("Predicted")
ax.set_ylabel("Actual")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig(cm_path, dpi=150)
plt.close(fig)
print("Saved confusion matrix to:", cm_path)

# Display inline if possible
try:
    display(Image(cm_path))
except Exception:
    print("Unable to display inline image; open file at:", cm_path)

# ------------- Append a compact timestamped summary to PROJECT_DIR/test_summaries.csv -------------
out_row = {
    "timestamp": pd.Timestamp.now().isoformat(),
    "test_loss": float(test_loss) if not np.isnan(test_loss) else "",
    "test_accuracy": float(test_acc) if not np.isnan(test_acc) else "",
    "num_test_samples": int(len(y_true)),
    "num_classes": int(len(class_names)),
    "micro_f1": float(micro_f1),
    "macro_f1": float(macro_f1),
    "weighted_f1": float(weighted_f1),
    "run_dir": RUN_DIR,
    "model_path": MODEL_PATH
}

summaries_path = os.path.join(PROJECT_DIR, "test_summaries.csv")
if not os.path.exists(summaries_path):
    pd.DataFrame([out_row]).to_csv(summaries_path, index=False)
else:
    dfm = pd.read_csv(summaries_path)
    dfm = pd.concat([dfm, pd.DataFrame([out_row])], ignore_index=True)
    dfm.to_csv(summaries_path, index=False)
print("Appended test summary to:", summaries_path)

print("\nDone — per-class CSV, confusion matrix PNG and test summary saved to:", OUT_DIR)
```

```python
# === Training overview plot (epoch-level preferred, fallback to PPO rollout log) ===
# Paste & run in the same Colab runtime. Adjust PROJECT_DIR / RUN_DIR / LOG_CSV if needed.

import os, glob, math, json
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display
import numpy as np

# ------------- user edit if needed -------------
RUN_DIR = None
PROJECT_DIR = "/content/drive/MyDrive/RL_CNN_Research/mobilenetv2_ppo_full"
# If you want to point directly to a log file, set LOG_CSV below; otherwise leave None to auto-find.
LOG_CSV = None
OUT_PNG = None
# -----------------------------------------------

def find_latest_run(project_dir):
    if not os.path.exists(project_dir):
        return None
    runs = [p for p in glob.glob(os.path.join(project_dir, "run_*")) if os.path.isdir(p)]
    if not runs:
        return None
    runs.sort(key=os.path.getmtime, reverse=True)
    return runs[0]

if RUN_DIR is None:
    RUN_DIR = find_latest_run(PROJECT_DIR)
    if RUN_DIR is None:
        raise FileNotFoundError("No run directory found. Set RUN_DIR or check PROJECT_DIR.")
print("Using RUN_DIR =", RUN_DIR)

# candidate files (epoch-level file optional; fallback to PPO rollout log)
epoch_hist = os.path.join(RUN_DIR, "per_epoch_history.csv")                 # optional epoch-level history
perclass_epoch = os.path.join(RUN_DIR, "per_epoch_per_class_metrics.csv")  # optional
log_csv_default = os.path.join(RUN_DIR, "ppo_onpolicy_log.csv")            # patched script default

if LOG_CSV is None:
    LOG_CSV = log_csv_default

if OUT_PNG is None:
    OUT_PNG = os.path.join(RUN_DIR, "training_overview.png")

print("Searching for files:")
print("  per-epoch history:", epoch_hist, "->", "FOUND" if os.path.exists(epoch_hist) else "MISSING")
print("  rollout log:", LOG_CSV, "->", "FOUND" if os.path.exists(LOG_CSV) else "MISSING")

# ---------- Try epoch-level first ----------
if os.path.exists(epoch_hist):
    df = pd.read_csv(epoch_hist)
    if 'epoch' not in df.columns:
        df = df.reset_index().rename(columns={'index':'epoch'})
        df['epoch'] = df['epoch'] + 1
    epochs = df['epoch'].astype(int).tolist()

    # ensure columns exist (fill with NaN if missing)
    for c in ['train_acc','val_acc','train_loss','val_loss','lr_before','action_mult','policy_loss','value_loss','entropy']:
        if c not in df.columns:
            df[c] = np.nan

    fig, axes = plt.subplots(3,1, figsize=(12,14), constrained_layout=True)

    # 1) val acc & val loss
    ax = axes[0]
    ax.plot(epochs, df['val_acc'], marker='o', label='val_acc')
    if not df['train_acc'].isna().all():
        ax.plot(epochs, df['train_acc'], marker='x', label='train_acc')
    ax.set_title("Validation & Train Accuracy")
    ax.set_xlabel("Epoch"); ax.set_ylabel("Accuracy"); ax.grid(True); ax.legend(loc='best')

    # val_loss on twin axis
    ax2 = ax.twinx()
    ax2.plot(epochs, df['val_loss'], linestyle='--', marker='s', label='val_loss', alpha=0.8)
    if not df['train_loss'].isna().all():
        ax2.plot(epochs, df['train_loss'], linestyle=':', marker='d', label='train_loss', alpha=0.6)
    ax2.set_ylabel("Loss")
    # combine legends
    lines = ax.get_lines() + ax2.get_lines()
    labels = [l.get_label() for l in lines]
    ax.legend(lines, labels, loc='upper right')

    # 2) LR & action multiplier
    ax = axes[1]
    ax.plot(epochs, df['lr_before'], marker='o', label='lr_before')
    if not df['action_mult'].isna().all():
        lr_after = df['lr_before'].values * df['action_mult'].values
        ax.scatter(epochs, lr_after, marker='x', label='lr_after')
    ax.set_yscale('log')
    ax.set_xlabel("Epoch"); ax.set_ylabel("Learning Rate (log)"); ax.grid(True); ax.legend(loc='upper left')

    # 3) policy/value/entropy losses
    ax = axes[2]
    plotted = False
    if not df['policy_loss'].isna().all():
        ax.plot(epochs, df['policy_loss'], marker='o', label='policy_loss'); plotted = True
    if not df['value_loss'].isna().all():
        ax.plot(epochs, df['value_loss'], marker='x', label='value_loss'); plotted = True
    if not df['entropy'].isna().all():
        ax.plot(epochs, df['entropy'], marker='s', label='entropy'); plotted = True
    if not plotted:
        ax.text(0.5,0.5,"No policy/value/entropy columns in epoch history", ha='center', va='center', transform=ax.transAxes)
    else:
        ax.set_xlabel("Epoch"); ax.set_ylabel("Loss/Entropy"); ax.grid(True); ax.legend(loc='best')

    plt.suptitle("Training overview (epoch-level)", fontsize=16, y=1.02)
    plt.savefig(OUT_PNG, dpi=150, bbox_inches='tight')
    plt.close(fig)
    print("Saved overview plot to:", OUT_PNG)
    try:
        display(Image(OUT_PNG))
    except Exception as e:
        print("Couldn't display inline image. Open file at:", OUT_PNG, "Error:", e)

# ---------- Fallback: use rollout-level PPO log ----------
elif os.path.exists(LOG_CSV):
    df = pd.read_csv(LOG_CSV)
    if df.empty:
        raise RuntimeError("Log CSV is empty.")
    # ordering by update_idx if available
    if 'update_idx' in df.columns:
        df = df.sort_values('update_idx').reset_index(drop=True)
        x = df['update_idx'].astype(int).tolist()
    else:
        df = df.reset_index().rename(columns={'index':'update_idx'})
        x = df['update_idx'].tolist()

    # ensure columns exist
    for c in ['mean_val_acc','lr','policy_loss','value_loss','entropy','chosen_actions','clip_eps','entropy_coef','kl']:
        if c not in df.columns:
            df[c] = np.nan

    fig, axes = plt.subplots(3,1, figsize=(12,14), constrained_layout=True)

    # 1) mean_val_acc (per-rollout)
    ax = axes[0]
    ax.plot(x, df['mean_val_acc'], marker='o', label='mean_val_acc (per-rollout)')
    ax.set_title("Rollout mean validation accuracy")
    ax.set_xlabel("rollout update_idx"); ax.set_ylabel("mean_val_acc"); ax.grid(True); ax.legend(loc='best')

    # 2) LR & chosen_actions (attempt to parse)
    ax = axes[1]
    ax.plot(x, df['lr'], marker='o', label='lr')
    parsed_actions = []
    for v in df['chosen_actions'].fillna('[]'):
        try:
            pv = json.loads(v) if isinstance(v, str) else v
            if isinstance(pv, (list,tuple)) and len(pv)>0:
                parsed_actions.append(pv[0])
            else:
                parsed_actions.append(np.nan)
        except Exception:
            try:
                s = str(v).strip()
                s = s.strip('[]')
                parsed_actions.append(float(s.split(',')[0]) if s else np.nan)
            except Exception:
                parsed_actions.append(np.nan)
    if any(~pd.isna(parsed_actions)):
        ax2 = ax.twinx()
        ax2.plot(x, parsed_actions, marker='x', linestyle='--', label='first_chosen_action(mult)')
        ax2.set_ylabel("action mult")
        ax2.legend(loc='upper right')
    ax.set_xlabel("rollout"); ax.set_ylabel("lr"); ax.grid(True); ax.legend(loc='upper left')

    # 3) policy/value/entropy
    ax = axes[2]
    plotted = False
    if not df['policy_loss'].isna().all():
        ax.plot(x, df['policy_loss'], marker='o', label='policy_loss'); plotted = True
    if not df['value_loss'].isna().all():
        ax.plot(x, df['value_loss'], marker='x', label='value_loss'); plotted = True
    if not df['entropy'].isna().all():
        ax.plot(x, df['entropy'], marker='s', label='entropy'); plotted = True
    if not plotted:
        ax.text(0.5,0.5,"No policy/value/entropy columns in log CSV", ha='center', va='center', transform=ax.transAxes)
    else:
        ax.set_xlabel("rollout"); ax.set_ylabel("loss/entropy"); ax.grid(True); ax.legend(loc='best')

    plt.suptitle("Training overview (rollout-level)", fontsize=16, y=1.02)
    plt.savefig(OUT_PNG, dpi=150, bbox_inches='tight')
    plt.close(fig)
    print("Saved overview plot to:", OUT_PNG)
    try:
        display(Image(OUT_PNG))
    except Exception as e:
        print("Couldn't display inline image. Open file at:", OUT_PNG, "Error:", e)

else:
    raise FileNotFoundError("No epoch history nor rollout log CSV found. Check run directory.")
```
