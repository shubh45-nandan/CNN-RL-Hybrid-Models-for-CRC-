```python

# -------------------- Colab Drive mount (robust) --------------------
MOUNT_POINT = '/content/drive'
try:
    # try a robust mount sequence that tolerates leftover files / earlier mounts
    from google.colab import drive
    try:
        drive.mount(MOUNT_POINT, force_remount=True)
    except Exception as e:
        # fallback: if the mountpoint contains stray files, remove and recreate
        import os, shutil
        try:
            if os.path.exists(MOUNT_POINT):
                # only attempt to remove when it's not an active mount; best-effort
                try:
                    shutil.rmtree(MOUNT_POINT)
                except Exception:
                    # if removal fails, continue and try mount anyway
                    pass
            os.makedirs(MOUNT_POINT, exist_ok=True)
            drive.mount(MOUNT_POINT, force_remount=True)
        except Exception as e2:
            # final fallback: raise helpful error
            raise RuntimeError(f"Drive mount failed: {e} / {e2}")
except Exception as e:
    # Not running in Colab (or import failed) — warn but continue
    print("Warning: google.colab.drive not available. Are you running this in Colab?", e)

# -------------------- imports --------------------
import os, glob, shutil, random, zipfile, math, json, datetime, itertools, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, optimizers, regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.applications.densenet import preprocess_input
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
import seaborn as sns

print("✅ Libraries ready, TF version:", tf.__version__)

# -------------------- USER EDIT --------------------
ZIP_IN_DRIVE = "/content/drive/MyDrive/archive.zip"   # <<-- edit to your zip
PROJECT_DIR  = "/content/drive/MyDrive/RL_CNN_Research/densenet_sac_exact"
os.makedirs(PROJECT_DIR, exist_ok=True)
RESULTS_MASTER = os.path.join(PROJECT_DIR, "results_master.csv")
# ---------------------------------------------------

# ------------- quick config -------------
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)
VAL_RATIO = 0.20; TEST_RATIO = 0.15
IMG_SIZE = (224,224); BATCH_SIZE = 32
EXTS = (".jpg",".jpeg",".png",".bmp",".tif",".tiff")

# unzip dataset (colab-friendly)
EXTRACT_DIR = "/content/Extracted_Kather_SAC_EXACT"
if os.path.exists(EXTRACT_DIR): shutil.rmtree(EXTRACT_DIR)
os.makedirs(EXTRACT_DIR, exist_ok=True)

if not os.path.exists(ZIP_IN_DRIVE):
    raise FileNotFoundError(f"ZIP not found at {ZIP_IN_DRIVE}. Did you set ZIP_IN_DRIVE correctly and mount Drive?")
print("Unzipping... (this may take a while for big zips)")
with zipfile.ZipFile(ZIP_IN_DRIVE, 'r') as zf:
    zf.extractall(EXTRACT_DIR)
print("Unzipped.")

def list_images_recursive(folder):
    out=[]
    for r,_,files in os.walk(folder):
        for f in files:
            if os.path.splitext(f)[1].lower() in EXTS:
                out.append(os.path.join(r,f))
    return sorted(out)

def find_dataset_root(base):
    candidates=[]
    for root, dirs, _ in os.walk(base):
        if not dirs: continue
        score=0
        for d in dirs:
            dpath=os.path.join(root,d)
            for f in os.listdir(dpath):
                if os.path.splitext(f)[1].lower() in EXTS:
                    score+=1; break
        if score>0: candidates.append((root,score))
    if candidates:
        candidates.sort(key=lambda x:x[1], reverse=True); return candidates[0][0]
    return base

DATA_ROOT = find_dataset_root(EXTRACT_DIR)
print("DATA_ROOT:", DATA_ROOT)

# map classes
class_to_files={}
for entry in sorted(os.listdir(DATA_ROOT)):
    p=os.path.join(DATA_ROOT, entry)
    if os.path.isdir(p):
        imgs=list_images_recursive(p)
        if imgs: class_to_files[entry]=imgs
if not class_to_files:
    # fallback scan deeper
    for r,dirs,files in os.walk(DATA_ROOT):
        for d in dirs:
            p=os.path.join(r,d)
            imgs=list_images_recursive(p)
            if imgs: class_to_files[d]=imgs

available_classes = sorted(class_to_files.keys())
print("Classes and counts:", [(c,len(class_to_files[c])) for c in available_classes])

# split
SPLIT_ROOT = "/content/kather_split_densenet_sac_exact"
if os.path.exists(SPLIT_ROOT): shutil.rmtree(SPLIT_ROOT)
TRAIN_DIR = os.path.join(SPLIT_ROOT,"train"); VAL_DIR = os.path.join(SPLIT_ROOT,"val"); TEST_DIR = os.path.join(SPLIT_ROOT,"test")
os.makedirs(TRAIN_DIR, exist_ok=True); os.makedirs(VAL_DIR, exist_ok=True); os.makedirs(TEST_DIR, exist_ok=True)

def allocate_counts(n,val_ratio=VAL_RATIO,test_ratio=TEST_RATIO):
    if n<=0: return (0,0,0)
    if n==1: return (1,0,0)
    if n==2: return (1,0,1)
    test_n=max(1,int(round(n*test_ratio))); val_n=max(1,int(round(n*val_ratio)))
    if test_n+val_n>=n: val_n=1; test_n=1
    train_n=n-val_n-test_n; return (train_n,val_n,test_n)

rng=np.random.default_rng(SEED)
tot={"train":0,"val":0,"test":0}
for cls, files in class_to_files.items():
    files = sorted(files); n=len(files); idx=np.arange(n); rng.shuffle(idx)
    tr_n, v_n, te_n = allocate_counts(n)
    tr_idx=set(idx[:tr_n]); v_idx=set(idx[tr_n:tr_n+v_n]); te_idx=set(idx[tr_n+v_n:tr_n+v_n+te_n])
    os.makedirs(os.path.join(TRAIN_DIR,cls), exist_ok=True)
    os.makedirs(os.path.join(VAL_DIR,cls), exist_ok=True)
    os.makedirs(os.path.join(TEST_DIR,cls), exist_ok=True)
    for i in idx:
        src=files[i]; basename=os.path.basename(src)
        if i in tr_idx: dst=os.path.join(TRAIN_DIR,cls,basename)
        elif i in v_idx: dst=os.path.join(VAL_DIR,cls,basename)
        else: dst=os.path.join(TEST_DIR,cls,basename)
        shutil.copy2(src,dst)
    tot['train']+=len(tr_idx); tot['val']+=len(v_idx); tot['test']+=len(te_idx)
print("Split done:", tot)

# datagens
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input,
                                   rotation_range=20, width_shift_range=0.1, height_shift_range=0.1,
                                   zoom_range=0.15, horizontal_flip=True)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_directory(TRAIN_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=True)
val_gen   = val_datagen.flow_from_directory(VAL_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False)
test_gen  = val_datagen.flow_from_directory(TEST_DIR, target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='categorical', shuffle=False)
class_names = list(train_gen.class_indices.keys()); num_classes=len(class_names)
print("Detected classes:", class_names)

# ------------- build model -------------
base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=IMG_SIZE+(3,))
base_model.trainable=False
inputs = keras.Input(shape=IMG_SIZE+(3,))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation='softmax')(x)
model = keras.Model(inputs, outputs)
INITIAL_LR = 1e-4
model.compile(optimizer=optimizers.Adam(learning_rate=INITIAL_LR), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# -------------------- SAC exact implementation --------------------
STATE_DIM = 3
ACTION_LOW, ACTION_HIGH = 0.5, 1.5
REPLAY_CAPACITY = 20000
BATCH_SAC = 64
SAC_UPDATES_PER_EPOCH = 64
GAMMA = 0.99
TAU = 0.005
POLICY_LR = 3e-4
Q_LR = 3e-4
ALPHA_LR = 3e-4
ACTION_DIM = 1
TARGET_ENTROPY = -float(ACTION_DIM)  # recommended heuristic
LOG_EPS = 1e-6
GRAD_CLIP_NORM = 1.0    # gradient clipping for stability
REPLAY_WARMUP = 12      # slightly larger warmup for small datasets

# ---------- ReplayBuffer (fixed __init__ & sample) ----------
class ReplayBuffer:
    def __init__(self, cap=REPLAY_CAPACITY):
        self.cap = int(cap)
        self.ptr = 0
        self.size = 0
        self.s = np.zeros((self.cap, STATE_DIM), dtype=np.float32)
        self.a = np.zeros((self.cap, 1), dtype=np.float32)
        self.r = np.zeros((self.cap, 1), dtype=np.float32)
        self.ns = np.zeros((self.cap, STATE_DIM), dtype=np.float32)
        self.d = np.zeros((self.cap, 1), dtype=np.float32)

    def push(self, s, a, r, ns, d):
        idx = self.ptr % self.cap
        self.s[idx] = s
        self.a[idx] = a
        self.r[idx] = r
        self.ns[idx] = ns
        self.d[idx] = d
        self.ptr += 1
        self.size = min(self.size + 1, self.cap)

    def sample(self, n):
        if self.size == 0:
            raise ValueError("Replay buffer is empty.")
        if self.size < n:
            # sample with replacement to fill batch if not enough
            idx = np.random.randint(0, self.size, size=n)
        else:
            idx = np.random.choice(self.size, size=n, replace=False)
        return dict(s=self.s[idx], a=self.a[idx], r=self.r[idx], ns=self.ns[idx], d=self.d[idx])

# ---------- simple MLP helper ----------
def mlp(in_dim, hidden=[256,256], out=1):
    inp = keras.Input(shape=(in_dim,))
    x = inp
    for h in hidden:
        x = layers.Dense(h, activation='relu')(x)
    outp = layers.Dense(out)(x)
    return keras.Model(inp, outp)

# ---------- Tanh-Gaussian policy (fixed __init__, TF-safe logs) ----------
class TanhGaussianPolicy:
    def __init__(self, state_dim, action_low, action_high, hidden=[256,256], lr=POLICY_LR):
        self.action_low = action_low
        self.action_high = action_high
        self.state_dim = state_dim
        # net outputs 2 units: mu and log_std (per action dimension)
        self.net = mlp(state_dim, hidden=hidden, out=2*ACTION_DIM)
        self.opt = optimizers.Adam(learning_rate=lr)

    def call(self, state):
        return self.net(state)

    def sample_action(self, state_np):
        """Return (action, pre_tanh, log_prob, mu, log_std) as numpy arrays for a batch of states."""
        state_tf = tf.convert_to_tensor(state_np.astype(np.float32))
        mu_log = self.net(state_tf)  # shape (N, 2*ACTION_DIM)
        mu = mu_log[:, :ACTION_DIM]
        log_std = mu_log[:, ACTION_DIM:ACTION_DIM*2]
        log_std = tf.clip_by_value(log_std, -20.0, 2.0)
        std = tf.exp(log_std)

        eps = tf.random.normal(shape=mu.shape)
        pre_tanh = mu + eps * std
        action_tanh = tf.tanh(pre_tanh)
        action = (action_tanh + 1.0) / 2.0 * (self.action_high - self.action_low) + self.action_low

        # Gaussian log prob with TF operations
        var = std ** 2
        const = tf.math.log(2.0 * tf.constant(math.pi, dtype=tf.float32))
        log_prob_gauss = -0.5 * (((pre_tanh - mu) ** 2) / (var + 1e-8) + 2.0 * log_std + const)
        log_prob_gauss = tf.reduce_sum(log_prob_gauss, axis=1, keepdims=True)
        log_det_jac = tf.reduce_sum(tf.math.log(1.0 - tf.tanh(pre_tanh) ** 2 + LOG_EPS), axis=1, keepdims=True)
        log_prob = log_prob_gauss - log_det_jac

        return action.numpy(), pre_tanh.numpy(), log_prob.numpy(), mu.numpy(), log_std.numpy()

    def get_vars(self):
        return self.net.trainable_variables

# ---------- Q networks ----------
q1 = mlp(STATE_DIM + ACTION_DIM, hidden=[256,256], out=1)
q2 = mlp(STATE_DIM + ACTION_DIM, hidden=[256,256], out=1)
q1_target = mlp(STATE_DIM + ACTION_DIM, hidden=[256,256], out=1)
q2_target = mlp(STATE_DIM + ACTION_DIM, hidden=[256,256], out=1)
q1_target.set_weights(q1.get_weights()); q2_target.set_weights(q2.get_weights())

policy = TanhGaussianPolicy(STATE_DIM, ACTION_LOW, ACTION_HIGH)
q1_opt = optimizers.Adam(learning_rate=Q_LR)
q2_opt = optimizers.Adam(learning_rate=Q_LR)
policy_opt = optimizers.Adam(learning_rate=POLICY_LR)
log_alpha = tf.Variable(0.0, dtype=tf.float32)
alpha_opt = optimizers.Adam(learning_rate=ALPHA_LR)

replay = ReplayBuffer(REPLAY_CAPACITY)

# helpers
def state_from(val_acc, prev_val_acc, lr):
    # val_acc, prev_val_acc in [0,1]; lr in [1e-6, 1e-3]
    delta = float(val_acc - prev_val_acc)
    lmin, lmax = math.log10(1e-6), math.log10(1e-3)
    lrn = (math.log10(max(lr,1e-12)) - lmin) / (lmax - lmin + 1e-8)
    return np.array([float(val_acc), float(delta), float(lrn)], dtype=np.float32)

def soft_update(target, source, tau=TAU):
    wt = target.get_weights(); ws = source.get_weights()
    new = [a*(1.0-tau)+b*tau for a,b in zip(wt,ws)]
    target.set_weights(new)

# TF-safe gaussian log prob used during SAC updates
def gaussian_log_prob(pre_tanh, mu, log_std):
    std = tf.exp(log_std)
    var = std ** 2
    const = tf.math.log(2.0 * tf.constant(math.pi, dtype=tf.float32))
    log_prob = -0.5 * (((pre_tanh - mu) ** 2) / (var + 1e-8) + 2.0 * log_std + const)
    return tf.reduce_sum(log_prob, axis=1, keepdims=True)

mse = tf.keras.losses.MeanSquaredError()

# SAC batch update
def sac_batch_update(batch):
    s = tf.convert_to_tensor(batch['s'], dtype=tf.float32)        # (B,STATE_DIM)
    a = tf.convert_to_tensor(batch['a'], dtype=tf.float32)        # (B,1)
    r = tf.convert_to_tensor(batch['r'], dtype=tf.float32)        # (B,1)
    ns = tf.convert_to_tensor(batch['ns'], dtype=tf.float32)      # (B,STATE_DIM)
    d = tf.convert_to_tensor(batch['d'], dtype=tf.float32)        # (B,1)

    # ---------- sample next action and log prob (on ns) ----------
    mu_log_next = policy.net(ns)
    mu_next = mu_log_next[:, :ACTION_DIM]; logstd_next = tf.clip_by_value(mu_log_next[:, ACTION_DIM:ACTION_DIM*2], -20.0, 2.0)
    std_next = tf.exp(logstd_next)
    eps_next = tf.random.normal(shape=mu_next.shape)
    pre_tanh_next = mu_next + eps_next * std_next
    action_tanh_next = tf.tanh(pre_tanh_next)
    a_next = (action_tanh_next + 1.0) / 2.0 * (ACTION_HIGH - ACTION_LOW) + ACTION_LOW

    logp_gauss_next = gaussian_log_prob(pre_tanh_next, mu_next, logstd_next)
    log_det_next = tf.reduce_sum(tf.math.log(1.0 - tf.tanh(pre_tanh_next)**2 + LOG_EPS), axis=1, keepdims=True)
    logp_next = logp_gauss_next - log_det_next  # shape (B,1)

    alpha = tf.exp(log_alpha)

    # ---------- target Q ----------
    q_input_next = tf.concat([ns, a_next], axis=1)
    q1_t = q1_target(q_input_next); q2_t = q2_target(q_input_next)
    q_target_min = tf.minimum(q1_t, q2_t)
    y = r + (1.0 - d) * GAMMA * (q_target_min - alpha * logp_next)

    # ---------- update Q networks ----------
    with tf.GradientTape() as tape1:
        q_input = tf.concat([s, a], axis=1)
        q1_pred = q1(q_input)
        q1_loss = tf.reduce_mean(0.5 * tf.square(y - q1_pred))
    grads1 = tape1.gradient(q1_loss, q1.trainable_variables)
    grads1, _ = tf.clip_by_global_norm(grads1, GRAD_CLIP_NORM)
    q1_opt.apply_gradients(zip(grads1, q1.trainable_variables))

    with tf.GradientTape() as tape2:
        q2_pred = q2(q_input)
        q2_loss = tf.reduce_mean(0.5 * tf.square(y - q2_pred))
    grads2 = tape2.gradient(q2_loss, q2.trainable_variables)
    grads2, _ = tf.clip_by_global_norm(grads2, GRAD_CLIP_NORM)
    q2_opt.apply_gradients(zip(grads2, q2.trainable_variables))

    # ---------- policy update ----------
    with tf.GradientTape() as tape_p:
        mu_log = policy.net(s)
        mu = mu_log[:, :ACTION_DIM]; logstd = tf.clip_by_value(mu_log[:, ACTION_DIM:ACTION_DIM*2], -20.0, 2.0)
        std = tf.exp(logstd)
        eps = tf.random.normal(shape=mu.shape)
        pre_tanh = mu + eps * std
        action_tanh = tf.tanh(pre_tanh)
        a_samp = (action_tanh + 1.0) / 2.0 * (ACTION_HIGH - ACTION_LOW) + ACTION_LOW

        logp_gauss = gaussian_log_prob(pre_tanh, mu, logstd)
        log_det = tf.reduce_sum(tf.math.log(1.0 - tf.tanh(pre_tanh)**2 + LOG_EPS), axis=1, keepdims=True)
        logp = logp_gauss - log_det

        q_input_for_policy = tf.concat([s, a_samp], axis=1)
        q1_pi = q1(q_input_for_policy); q2_pi = q2(q_input_for_policy)
        q_pi = tf.minimum(q1_pi, q2_pi)

        policy_loss = tf.reduce_mean(alpha * logp - q_pi)   # minimize (alpha*logp - Q)
    grads_p = tape_p.gradient(policy_loss, policy.get_vars())
    grads_p, _ = tf.clip_by_global_norm(grads_p, GRAD_CLIP_NORM)
    policy_opt.apply_gradients(zip(grads_p, policy.get_vars()))

    # ---------- alpha (temperature) update to match target entropy ----------
    with tf.GradientTape() as tape_a:
        target = tf.constant(TARGET_ENTROPY, dtype=tf.float32)
        # note: 'logp' from policy sampling above is shape (B,1)
        alpha_loss = -tf.reduce_mean(log_alpha * (logp + target))
    grads_a = tape_a.gradient(alpha_loss, [log_alpha])
    alpha_opt.apply_gradients(zip(grads_a, [log_alpha]))

    return float(q1_loss.numpy()), float(q2_loss.numpy()), float(policy_loss.numpy()), float(alpha_loss.numpy())

# -------------------- Training loop helpers --------------------
best_val_acc = -1.0
current_lr = INITIAL_LR
prev_val_acc = 0.0

LOG_CSV = os.path.join(PROJECT_DIR, "sac_exact_log.csv")
if not os.path.exists(LOG_CSV):
    pd.DataFrame(columns=['timestamp','phase','epoch','val_acc','val_loss','lr','action','reward','q1_loss','q2_loss','policy_loss','alpha_loss','alpha']).to_csv(LOG_CSV,index=False)

CKPT_DIR = os.path.join(PROJECT_DIR, "checkpoints"); os.makedirs(CKPT_DIR, exist_ok=True)
BEST_MODEL_PATH = os.path.join(PROJECT_DIR, "densenet121_colorectal_SAC_exact_best.keras")
FINAL_MODEL_PATH = os.path.join(PROJECT_DIR, "densenet121_colorectal_SAC_exact_final.keras")

# Small helper: rotate old checkpoints and keep last N
def save_checkpoint(model, phase, epoch_idx, keep_last=3):
    fname = os.path.join(CKPT_DIR, f"{phase}_epoch{epoch_idx}.keras")
    model.save(fname)
    # rotate older ckpts matching phase
    all_ck = sorted([p for p in glob.glob(os.path.join(CKPT_DIR, f"{phase}_epoch*.keras"))], key=os.path.getmtime, reverse=True)
    for old in all_ck[keep_last:]:
        try: os.remove(old)
        except: pass
    return fname

def run_epoch_and_sac(phase, epoch_idx, epochs_total, use_random_action=False):
    global current_lr, prev_val_acc, best_val_acc
    # Train for 1 epoch (explicit steps to avoid generator ambiguity)
    h = model.fit(train_gen, validation_data=val_gen, epochs=1, verbose=1,
                  steps_per_epoch=len(train_gen), validation_steps=len(val_gen))
    val_acc = float(h.history.get('val_accuracy', [0.0])[-1])
    val_loss = float(h.history.get('val_loss', [0.0])[-1])

    # state
    s = state_from(val_acc, prev_val_acc, current_lr)

    # sample action (policy) OR random for warmup
    if use_random_action:
        a_scalar = float(np.random.uniform(ACTION_LOW, ACTION_HIGH))
        # for logging and storing, create dummy pre_tanh/logp etc
        pre_tanh_np = np.array([[0.0]]); logp_np = np.array([[0.0]]); mu_np = np.array([[0.0]]); logstd_np = np.array([[0.0]])
    else:
        a_np, pre_tanh_np, logp_np, mu_np, logstd_np = policy.sample_action(np.expand_dims(s, axis=0))
        a_scalar = float(np.clip(a_np.ravel()[0], ACTION_LOW, ACTION_HIGH))

    # next lr
    new_lr = float(np.clip(current_lr * a_scalar, 1e-6, 1e-3))
    # reward choice: use val_acc (optionally could use smoothed metric)
    reward = float(val_acc)
    ns = state_from(val_acc, prev_val_acc, new_lr)
    done = 0.0

    # store in replay: shapes must match (STATE_DIM,) and (1,)
    replay.push(s, np.array([[a_scalar]], dtype=np.float32), np.array([[reward]], dtype=np.float32), ns, np.array([[done]], dtype=np.float32))

    # apply lr to optimizer with safe fallbacks
    try:
        # prefer assign
        model.optimizer.learning_rate.assign(new_lr)
    except Exception:
        try:
            tf.keras.backend.set_value(model.optimizer.learning_rate, new_lr)
        except Exception:
            # fallback: recompile (rare)
            model.compile(optimizer=optimizers.Adam(learning_rate=new_lr), loss='categorical_crossentropy', metrics=['accuracy'])
    current_lr = new_lr

    # SAC updates
    q1_l = q2_l = p_l = a_l = 0.0
    if replay.size >= max(BATCH_SAC, REPLAY_WARMUP * len(train_gen)):  # ensure buffer has enough diversity
        for _ in range(SAC_UPDATES_PER_EPOCH):
            batch = replay.sample(BATCH_SAC)
            q1_loss, q2_loss, policy_loss, alpha_loss = sac_batch_update(batch)
            q1_l += q1_loss; q2_l += q2_loss; p_l += policy_loss; a_l+= alpha_loss
        n = SAC_UPDATES_PER_EPOCH
        q1_l/=n; q2_l/=n; p_l/=n; a_l/=n
        soft_update(q1_target, q1, TAU); soft_update(q2_target, q2, TAU)

    # logging to CSV (use pd.concat instead of deprecated append)
    row = dict(timestamp=str(datetime.datetime.now()), phase=phase, epoch=epoch_idx, val_acc=val_acc, val_loss=val_loss,
               lr=current_lr, action=a_scalar, reward=reward, q1_loss=float(q1_l), q2_loss=float(q2_l), policy_loss=float(p_l), alpha_loss=float(a_l), alpha=float(tf.exp(log_alpha).numpy()))
    df = pd.read_csv(LOG_CSV)
    df = pd.concat([df, pd.DataFrame([row])], ignore_index=True)
    df.to_csv(LOG_CSV, index=False)

    # checkpoints
    saved_path = save_checkpoint(model, phase, epoch_idx, keep_last=3)
    saved_best=False
    if val_acc > best_val_acc:
        best_val_acc = val_acc; model.save(BEST_MODEL_PATH); saved_best=True

    print(f"[{phase}] Epoch {epoch_idx}/{epochs_total}: val_acc={val_acc:.4f}, new_lr={current_lr:.6f}, best_val={best_val_acc:.4f}, best_saved={saved_best}")
    prev_val_acc = val_acc

# -------------------- Training phases --------------------
PHASE1_EPOCHS = 15
PHASE2_EPOCHS = 25

print("=== Phase 1: frozen training (SAC controls LR) ===")
for e in range(1, PHASE1_EPOCHS+1):
    # allow random actions for first few epochs to warm replay buffer
    use_rand = (e <= REPLAY_WARMUP)
    run_epoch_and_sac("frozen", e, PHASE1_EPOCHS, use_random_action=use_rand)

print("=== Phase 2: fine-tune (unfreeze top layers) ===")
base_model.trainable=True
# freeze most layers, unfreeze top 50 (tuneable)
for layer in base_model.layers[:-50]:
    layer.trainable=False
for layer in base_model.layers[-50:]:
    layer.trainable=True

# recompile with current lr
model.compile(optimizer=optimizers.Adam(learning_rate=current_lr), loss='categorical_crossentropy', metrics=['accuracy'])

for e in range(1, PHASE2_EPOCHS+1):
    run_epoch_and_sac("finetune", e, PHASE2_EPOCHS, use_random_action=False)

# -------------------- Final evaluation, per-class logging, append to master --------------------
print("Final evaluation on test set...")
test_res = model.evaluate(test_gen, verbose=1, steps=len(test_gen))
test_acc = float(test_res[1]) if len(test_res)>1 else float(test_res[0])
y_true = test_gen.classes
y_pred = np.argmax(model.predict(test_gen, verbose=0, steps=len(test_gen)), axis=1)

# classification report + per-class csv
report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
report_df = pd.DataFrame(report).transpose()
per_class_csv = os.path.join(PROJECT_DIR, "per_class_report.csv")
report_df.to_csv(per_class_csv)
print("Saved per-class report:", per_class_csv)

# Save per-class counts
counts = pd.Series(y_true).value_counts().sort_index()
counts.index = [class_names[i] for i in counts.index]
counts.to_frame('count').to_csv(os.path.join(PROJECT_DIR, "per_class_counts.csv"))

# misclassification log (list of files mispredicted) + csv
misclassified_dir = os.path.join(PROJECT_DIR, "misclassified")
if os.path.exists(misclassified_dir): shutil.rmtree(misclassified_dir)
os.makedirs(misclassified_dir, exist_ok=True)
mis_rows=[]
# iterate over test_gen filenames to collect misclassified examples
filenames = test_gen.filepaths
for i, fp in enumerate(filenames):
    if i>=len(y_pred): break
    t = int(y_true[i]); p = int(y_pred[i])
    if p != t:
        mis_rows.append({'filepath': fp, 'true': class_names[t], 'pred': class_names[p]})
        dst = os.path.join(misclassified_dir, f"true_{class_names[t]}_pred_{class_names[p]}_{os.path.basename(fp)}")
        try:
            shutil.copy2(fp, dst)
        except Exception:
            # if the test file is inside zipped/virtual FS, best-effort
            pass
mis_csv = os.path.join(PROJECT_DIR, "misclassified_list.csv")
pd.DataFrame(mis_rows).to_csv(mis_csv, index=False)
print("Saved misclassified examples to:", misclassified_dir)
print("Misclassified list saved to:", mis_csv)

# confusion matrix plot
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10,8)); sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix — DenseNet121 + SAC (Exact)"); plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.show()

# append to master CSV (use pd.concat)
final_row = {
    'timestamp': datetime.datetime.now().isoformat(),
    'method': 'DenseNet121+SAC_exact',
    'test_accuracy': float(test_acc),
    'num_classes': int(num_classes),
    'best_val_accuracy': float(best_val_acc),
    'final_model_path': FINAL_MODEL_PATH,
    'best_model_path': BEST_MODEL_PATH,
    'log_csv': LOG_CSV,
    'per_class_report': per_class_csv,
    'misclassified_dir': misclassified_dir,
    'misclassified_list_csv': mis_csv
}
if not os.path.exists(RESULTS_MASTER):
    pd.DataFrame([final_row]).to_csv(RESULTS_MASTER, index=False)
else:
    dfm = pd.read_csv(RESULTS_MASTER)
    dfm = pd.concat([dfm, pd.DataFrame([final_row])], ignore_index=True)
    dfm.to_csv(RESULTS_MASTER, index=False)
print("Appended run summary to:", RESULTS_MASTER)

# save final model
model.save(FINAL_MODEL_PATH)
print("Saved final model to:", FINAL_MODEL_PATH)
print("Saved best model to:", BEST_MODEL_PATH)
print("Saved log to:", LOG_CSV)
print("Per-class report:", per_class_csv)
print("Per-class counts:", os.path.join(PROJECT_DIR, "per_class_counts.csv"))
print("Misclassified examples dir:", misclassified_dir)
```

```python
# Robust Drive mount helper (replace your drive.mount(...) line with this)
import os
from google.colab import drive

def safe_mount(mount_point='/content/drive'):
    # If the usual Drive path already exists, assume mounted and skip.
    if os.path.exists(os.path.join(mount_point, 'MyDrive')):
        print(f"Drive already mounted at {mount_point} (found MyDrive).")
        return
    try:
        drive.mount(mount_point, force_remount=False)
        print("Drive mounted.")
    except ValueError as e:
        # Mountpoint not empty / other ValueError
        print("drive.mount ValueError:", e)
        # If MyDrive exists now (possibly transient), treat as mounted
        if os.path.exists(os.path.join(mount_point, 'MyDrive')):
            print("Drive appears mounted after error — continuing.")
            return
        # Try flush/unmount (safe if previously mounted), then remount with force
        try:
            print("Attempting flush_and_unmount() then remount with force_remount=True...")
            drive.flush_and_unmount()
        except Exception as ex_flush:
            print("flush_and_unmount() failed or unnecessary:", ex_flush)
        # Final attempt with force_remount
        drive.mount(mount_point, force_remount=True)
        print("Drive mounted with force_remount=True.")

# call it
safe_mount('/content/drive')
```
